```python
import requests
import pandas as pd

API_KEY = "41a5eecb"
BASE_URL = "https://www.omdbapi.com/"

all_results = []

for page in range(1, 630):
    params = {
        "apikey": API_KEY,
        "s": "batman",
        "page": page
    }

    response = requests.get(BASE_URL, params=params)
    data = response.json()

    # Stop if no results are returned
    if "Search" not in data:
        print(f"Stopped at page {page}")
        break

    temp_df = pd.DataFrame(data["Search"])[["Title", "Year"]]
    all_results.append(temp_df)

# Combine all pages into one DataFrame
df = pd.concat(all_results, ignore_index=True)

print(df.head())
print(f"Total records: {len(df)}")
```
