## What is the Curse of Dimensionality?

As dimensions increase:
1) Data becomes sparse
2) Distance measures become less meaningful
3) Models need much more data to learn patterns
4) Computation cost increases

### Why is it a problem in ML?

1) Data Sparsity

**High-dimensional space is mostly empty**

**Hard to learn reliable patterns**

2) Distance Becomes Meaningless

**Algorithms like KNN, K-means, SVM rely on distance**

**In high dimensions, nearest ≈ farthest**

3) Overfitting

**More features → more noise**

**Model memorizes instead of generalizing**

4) Computational Cost

**Training time and memory increase exponentially**

## Solution:

**Dimensionality Reduction**
1) Feature Selection
2) Feature Extraction
3) Increase sample size
4) Regularization


   a) **Feature Selection**
   1) Forward selection
   2) Backward elimination
      
   b) **Feature Extraction**
   1) PCA
   2) LDA
   3) tsne [t-SNE (t-Distributed Stochastic Neighbor Embedding) is an unsupervised, non-linear dimensionality reduction method]
