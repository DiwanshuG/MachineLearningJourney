```python
import numpy as np
import plotly.graph_objects as go

# Sample dataset
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
n = len(x)

# Create range of m and b values
m_vals = np.linspace(-2, 4, 100)
b_vals = np.linspace(-2, 6, 100)

M, B = np.meshgrid(m_vals, b_vals)

# Compute Loss Surface (MSE)
loss = np.zeros_like(M)

for i in range(len(m_vals)):
    for j in range(len(b_vals)):
        y_pred = M[j, i] * x + B[j, i]
        loss[j, i] = np.mean((y - y_pred) ** 2)

# Create 3D surface plot
fig = go.Figure(data=[
    go.Surface(x=M, y=B, z=loss)
])

fig.update_layout(
    title='3D Loss Surface (MSE)',
    scene=dict(
        xaxis_title='Slope (m)',
        yaxis_title='Intercept (b)',
        zaxis_title='Loss'
    ),
    width=800,
    height=800
)

fig.show()
```
