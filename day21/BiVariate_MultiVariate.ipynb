## Pairplot: Iris Dataset Feature Relationships

This pairplot visualizes the pairwise relationships between numerical features in the Iris dataset, with points colored by species.

### What this plot shows
- Diagonal plots represent the distribution of each feature.
- Off-diagonal plots show scatterplots between feature pairs.
- Color (hue = species) distinguishes between the three Iris species:
  - Setosa
  - Versicolor
  - Virginica

### Features included
- sepal_length
- sepal_width
- petal_length
- petal_width

### Key Insights
- Setosa is clearly separable from the other species, especially in petal length and petal width.
- Versicolor and Virginica show partial overlap but can still be distinguished using petal-based features.
- Petal features provide better class separation than sepal features.

### Why Pairplot?
Pairplots offer a fast multivariate overview, helping identify:
- Feature correlations
- Natural clustering of classes
- Which features are most useful for classification

This visualization suggests that petal measurements are the most discriminative features for species classification.

```python
## What is the total / average number of passengers per year?
```

```python
yearly = flight.groupby("year")["passengers"].sum().reset_index()
```

```python
sns.lineplot(data=yearly,x="year",y="passengers")
plt.title("Total Airline Passenger per year")
plt.ylabel("Total Passenger")
plt.show()
```

```python
flight["year"].describe()
```

#### Goal : Show how passenger counts vary by Month and Year

This is:
1) Categorical (month)  Categorical (year)
2) Value = passengers (numeric)

```python
pivot_flights = flight.pivot_table(
    index="month",
    columns="year",
    values="passengers",
    observed=False
)
```

```python
sns.heatmap(
    pivot_flights,
    cmap="YlGnBu"
)

plt.title("Monthly Airline Passengers Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

```python
sns.clustermap(
    pivot_flights,
    cmap="YlGnBu"
)
```
