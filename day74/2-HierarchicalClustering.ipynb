# Hierarchical Clustering 

### why k-means is not always useful

k-means relies on distance to centroids and assumes:
- clusters are spherical  
- clusters are well separated  
- similar density  

fails in:
- crescent / moon datasets  
- spiral datasets  
- elongated / parallel clusters  
- noisy data  

reason:
k-means creates linear boundaries and cannot capture complex shapes

---

### hierarchical clustering

builds clusters in a tree-like structure (dendrogram)

types:
- agglomerative (bottom-up)  
- divisive (top-down)  

---

### agglomerative clustering (most used)

idea:
- start with each point as its own cluster  
- merge closest clusters step by step  
- continue until one cluster remains  

flow:

n clusters → n-1 → n-2 → ... → 1


---

### divisive clustering

- start with one cluster  
- split into smaller clusters  
- continue splitting  

flow:

1 → 2 → 3 → ... → n


less used because splitting correctly is harder than merging

---


after merging clusters → matrix is updated

---

### distance calculation

euclidean distance:

$$
d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$$

general form:

$$
d = \sqrt{\sum (x_i - y_i)^2}
$$

in practice:
- we often skip sqrt (faster, same result)

---

### linkage methods (very important)

defines how distance between clusters is computed

- single linkage → minimum distance  
- complete linkage → maximum distance  
- average linkage → mean distance  
- ward linkage → minimizes variance  

intuition:
- single → can form chains (good for complex shapes, bad with noise)  
- complete → compact clusters  
- average → balanced  
- ward → best default for spherical clusters  

---

### dendrogram

tree representation of merging process

- x-axis → data points  
- y-axis → distance (dissimilarity)  

---

### choosing number of clusters (trick)

- plot dendrogram  
- find longest vertical gap (no horizontal cuts)  
- draw horizontal line  
- count intersections  

that count = number of clusters  

note:
- vertical height = inter-cluster distance  
- larger height → clusters are less similar  

---

### hyperparameters (agglomerative)

- n_clusters → number of clusters  
- linkage → how clusters are merged  
- metric → distance measure  
- distance_threshold → stop merging based on distance  

---

### best practices

- always scale data (important for distance-based methods)  
- start with:
  - linkage = ward  
  - metric = euclidean  

- use dendrogram to decide clusters  
- visualize with PCA  

---

### when to use what

k-means:
- fast  
- good for spherical clusters  

hierarchical clustering:
- better for complex data  
- gives hierarchy (more insight)  
- useful when number of clusters is unknown  

---

### key takeaway

- k-means = simple but limited  
- hierarchical = flexible but slower  
- dendrogram = powerful tool for understanding clusters  

[Hyperparamters](https://scikit-learn.org/stable/auto_examples/cluster/plot_agglomerative_clustering_metrics.html)

### benefits of hierarchical clustering

1) works on a wide variety of datasets  
   it is widely applicable and does not assume a fixed cluster shape  

2) provides full hierarchical information  
   using a dendrogram, we can understand which points or clusters are closer to each other at every step  

### limitations of hierarchical clustering

1) not suitable for very large datasets  
   for large data (e.g., 10^6 points), computing and storing the proximity matrix requires very high memory  

2) computationally expensive  
   merging clusters step by step increases time complexity significantly
