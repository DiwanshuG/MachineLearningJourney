# Bagging = Bootstrapping + Aggregation

Bagging (Bootstrap Aggregating) is an ensemble learning method where:
- **Bootstrapping** creates different training datasets using random sampling with replacement.
- **Aggregation** combines predictions from multiple models to produce one final output.

## What is Bootstrapping?
Bootstrapping means creating many random samples from the original dataset.
- Suppose we have **10,000 rows**.
- We create sampled datasets like **D1, D2, D3, ..., D100**.
- Each sampled dataset is formed by random sampling **with replacement**.
- These sampled datasets are then used to train different models.

## What is Aggregation?
Aggregation means combining outputs of all trained models.
- For **classification**, we use **majority voting**.
- The class with the highest vote count becomes the final prediction.

## How Bagging Works (Step-by-Step)
Assume we need to perform a **classification** task on a dataset of **10k rows**.

1. Choose multiple base models: $M_1, M_2, M_3, \dots, M_{100}$.
2. Important rule: all base models should use the **same algorithm** (for example, all Decision Trees).
3. Create bootstrap samples:
- Sample **D1** and train $M_1$.
- Sample **D2** and train $M_2$.
- Continue similarly up to $D_{100}$ for $M_{100}$.
4. After training, a **new query point** arrives.
5. Send this query point to all 100 models.
6. Collect all predicted classes.
7. Apply **majority vote** to select the final class.

## Why Performance Improves
- Different bootstrap samples create model diversity (variation).
- Combining many model outputs reduces variance.
- Final performance is usually more stable and better than a single base model.

## Why Does Bagging Work? (Bias-Variance View)

### 1) Core Problem: Bias-Variance Tradeoff
- **Bias**: model cannot fit training data well (underfitting tendency).
- **Variance**: model is too sensitive to small data changes (overfitting tendency).
- We want **low bias + low variance**, but in practice they are often negatively correlated.

### 2) Typical ML Behavior
Most algorithms usually fall into one side:
- High bias, low variance
- Low bias, high variance

### 3) Bagging Strategy
Bagging starts with a base model that has:
- **Low bias**
- **High variance**

Example:
- Fully grown Decision Tree (`max_depth=None`)
- Such trees can overfit easily, so variance is high.

### 4) What Bootstrapping Fixes
During bagging:
- Each model trains on a different bootstrap sample.
- Outlier impact gets distributed across many samples/models.
- Model errors become less correlated.

Result:
- **Variance decreases** after aggregation.
- Bias remains low (base learner is still strong).

### 5) Final Outcome
Bagging turns a low-bias/high-variance learner into an ensemble with:
- **Low bias**
- **Lower variance**
- Better and more stable performance.

## Types of Bagging (Based on Sampling Strategy)

Bagging techniques can vary depending on how the data is sampled. Different sampling strategies lead to different variations of Bagging.

---

### 1. Pasting (Row Sampling without Replacement)

- In this approach, we perform row sampling without replacement  
- Each model is trained on a unique subset of the data  
- Unlike traditional Bagging, the same data point is not repeated  

This method is useful when we want diversity in training data without duplication.

---

### 2. Random Subspaces (Column Sampling)

- Instead of sampling rows, we sample features (columns)  
- Each model is trained on a subset of features  

This approach is commonly used in high-dimensional datasets such as image and text data.

---

### 3. Random Patches (Row and Column Sampling)

- Combines both row sampling and column sampling  
- Each model is trained on:
  - A subset of data points  
  - A subset of features  

This increases diversity among models and often improves generalization performance.

---

### Summary

| Method | Sampling Type |
|--------|--------------|
| Pasting | Row sampling (without replacement) |
| Random Subspaces | Column sampling |
| Random Patches | Row and column sampling |

---

These variations of Bagging introduce diversity in different ways, helping improve model stability and performance.
