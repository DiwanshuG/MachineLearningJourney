```python
import pandas as pd
from sklearn.preprocessing import Binarizer

df = pd.read_csv("train.csv", usecols=["Age", "Survived"]).dropna()

age_binarizer = Binarizer(threshold=18)

df["is_child"] = age_binarizer.fit_transform(df[["Age"]])

df[["Age", "is_child", "Survived"]].head(10)
```

```python
df = pd.read_csv("train.csv", usecols=["Age", "Fare", "Survived"]).dropna()

# Binarize Fare
fare_binarizer = Binarizer(threshold=30)
df["high_fare"] = fare_binarizer.fit_transform(df[["Fare"]])

# Inspect result
df[["Fare", "high_fare", "Survived"]].head(10)
```
