#### Independent Events & Mutually Exclusive Events 


##### Independent Events 

Two events are called independent if the occurrence of one event does not affect the probability of the other.

In simple terms:

P(A | B) = P(A)

This means knowing that B happened does not change the probability of A.

---

##### Mathematical Form

For independent events:

P(A ∩ B) = P(A) × P(B)

---

##### Example

Suppose:

- Event A = getting Head when a coin is tossed → P(A) = 1/2  
- Event B = getting 4 on a dice → P(B) = 1/6  

These two events are independent because the coin toss does not affect the dice roll.

Now:

P(A ∩ B) = (1/2) × (1/6) = 1/12

So, the probability of getting a Head and a 4 together is 1/12.

> Independent events are events that can occur together, but the occurrence of one event does not affect the probability of the other.



##### Independent Events

Let A and B be two events.

---

##### Conditional Probability

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}
$$

---

##### For Independent Events

$$
P(A \cap B) = P(A) \times P(B)
$$

---

##### Substituting in Conditional Probability

$$
P(A \mid B) = \frac{P(A)\times P(B)}{P(B)}
$$

---

##### Simplification

$$
P(A \mid B) = P(A)
$$

---

##### Conclusion

- $$P(A \cap B) = P(A)\times P(B)$$  
- $$P(A \mid B) = P(A)$$  

This shows that the occurrence of B does not affect the probability of A.

## Mutually Exclusive Events

Two events are called mutually exclusive if they cannot occur at the same time.

In simple terms:

If one event happens, the other cannot happen.

---

##### Mathematical Condition

$$
P(A \cap B) = 0
$$

---

##### Addition Rule

For mutually exclusive events:

$$
P(A \cup B) = P(A) + P(B)
$$

---

##### Example

When rolling a dice:

- A = getting an even number → {2, 4, 6}  
- B = getting an odd number → {1, 3, 5}  

A and B cannot occur together, so they are mutually exclusive.

$$
P(A \cap B) = 0
$$

---

##### 

Mutually exclusive events have no overlap, unlike independent events.

- Mutually Exclusive → cannot happen together  
- Independent → can happen together but do not affect each other
