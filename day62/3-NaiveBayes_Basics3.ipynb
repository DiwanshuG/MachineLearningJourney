##### Bayes Theorem

Bayes Theorem helps us find the probability of an event based on prior knowledge of related events.

In simple terms:

It tells us how to update probabilities when new information is available.

---

##### Formula

$$
P(A \mid B) = \frac{P(B \mid A)\, P(A)}{P(B)}
$$

---

##### Meaning of Terms

- $$P(A \mid B)$$ → Posterior (probability of A given B)  
- $$P(B \mid A)$$ → Likelihood (probability of B given A)  
- $$P(A)$$ → Prior (initial belief about A)  
- $$P(B)$$ → Evidence (total probability of B)  

---

##### Example (Intuition)

Suppose:

- A = person has a disease  
- B = test result is positive  

We want:

$$
P(\text{Disease} \mid \text{Positive})
$$

Bayes theorem helps us reverse the probability:

From:

$$
P(\text{Positive} \mid \text{Disease})
$$

To:

$$
P(\text{Disease} \mid \text{Positive})
$$

---



Bayes Theorem is used to reverse conditional probabilities.

It forms the foundation of Naive Bayes algorithm in Machine Learning.

##### Bayes Theorem - Derivation

Start with conditional probability:

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)} \quad (1)
$$

---

##### Property of Intersection

$$
A \cap B = B \cap A
$$

---

##### Another Conditional Form

$$
P(B \mid A) = \frac{P(B \cap A)}{P(A)} = \frac{P(A \cap B)}{P(A)} \quad (2)
$$

---

##### Rearranging (2)

$$
P(A \cap B) = P(B \mid A) \cdot P(A)
$$

---

##### Substitute in (1)

$$
P(A \mid B) = \frac{P(B \mid A)\cdot P(A)}{P(B)}
$$

---

##### Final Result (Bayes Theorem)

$$
P(A \mid B) = \frac{P(B \mid A)\, P(A)}{P(B)}
$$

### Example 

##### Bayes Theorem - Marker Factory Example

A factory manufactures markers using three machines:

- M1 → 20% production, 5% defect rate  
- M2 → 30% production, 3% defect rate  
- M3 → 50% production, 1% defect rate  

---

##### Define Events

- A1 = Marker is from M1  
- A2 = Marker is from M2  
- A3 = Marker is from M3  
- D = Marker is defective  

We need to find:

$$
P(A_3 \mid D)
$$

---

##### Step 1: Given Probabilities

$$
P(A_1)=0.2,\quad P(A_2)=0.3,\quad P(A_3)=0.5
$$

$$
P(D \mid A_1)=0.05,\quad P(D \mid A_2)=0.03,\quad P(D \mid A_3)=0.01
$$

---

##### Step 2: Total Probability of Defective Marker

$$
P(D) = P(D \mid A_1)P(A_1) + P(D \mid A_2)P(A_2) + P(D \mid A_3)P(A_3)
$$

$$
P(D) = (0.05 \times 0.2) + (0.03 \times 0.3) + (0.01 \times 0.5)
$$

$$
P(D) = 0.01 + 0.009 + 0.005 = 0.024
$$

---

##### Step 3: Apply Bayes Theorem

$$
P(A_3 \mid D) = \frac{P(D \mid A_3)\cdot P(A_3)}{P(D)}
$$

$$
P(A_3 \mid D) = \frac{0.01 \times 0.5}{0.024}
$$

$$
P(A_3 \mid D) = \frac{0.005}{0.024} \approx 0.2083
$$

---

##### Final Answer

$$
P(A_3 \mid D) \approx 0.2083
$$

---

#####

Even though M3 produces 50% of the markers, the probability that a defective marker comes from M3 is only about **20.83%**, because its defect rate is very low.
