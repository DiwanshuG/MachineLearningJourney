## ConditionalProbability


Conditional probability is the probability of an event occurring given that another event has already occurred.

It is written as:

$$
P(A | B) = P(A ∩ B) / P(B)
$$
This means we are only considering the cases where event B has already happened.

---

##### Example

Suppose:

- Total students = 100  
- Students who like Football (B) = 50  
- Students who like both Cricket and Football (A ∩ B) = 20  

Then:

P(Cricket | Football) = 20 / 50 = 0.4

So, if a student likes football, there is a 40% chance they also like cricket.

##### - Example (Two Dice)

When two dice are rolled together, the total number of possible outcomes is 36.

Let:

- A = getting a sum of 8  
- B = getting a sum greater than 5  

First, find outcomes:

- Sum = 8 → (2,6), (3,5), (4,4), (5,3), (6,2) → 5 outcomes  
- Sum > 5 → 30 outcomes  
- Common outcomes (A ∩ B) → same 5 outcomes  

Now:

P(A | B) = 5 / 30 = 1 / 6

So, given that the sum is greater than 5, the probability that the sum is 8 is 1/6.

##### example 2

Two dice are rolled.

Let:

- A = first die (D1) is 5  
- B = sum of two dice (D1 + D2) ≤ 10  

We need to find:

P(D1 = 5 | D1 + D2 ≤ 10)

---

##### Step 1: Total outcomes where sum ≤ 10

Total outcomes = 36

Exclude sums 11 and 12:

- Sum = 11 → (5,6), (6,5) → 2 outcomes  
- Sum = 12 → (6,6) → 1 outcome  

So,

Total outcomes where B is true = 36 - 3 = 33

---

##### Step 2: Outcomes where D1 = 5 and sum ≤ 10

Possible outcomes with D1 = 5:

(5,1), (5,2), (5,3), (5,4), (5,5), (5,6)

Now apply condition (sum ≤ 10):

- (5,1) → 6 
- (5,2) → 7   
- (5,3) → 8   
- (5,4) → 9   
- (5,5) → 10

    
- (5,6) → 11 ❌  

Valid outcomes = 5

---

##### Final Answer

P(D1 = 5 | D1 + D2 ≤ 10) = 5 / 33
