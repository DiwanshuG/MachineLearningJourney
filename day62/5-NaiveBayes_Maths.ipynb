##### Notation

Let:

$$
X = \{x_1, x_2, x_3, \dots, x_n\}
$$

- X represents the set of input features  

---

$$
C_k = \{c_1, c_2, c_3, \dots, c_k\}
$$

- \( C_k \) represents the **output (target) column**  
- Each \( c_i \) is a possible class label  

---

##### Meaning

- \( x_1, x_2, ..., x_n \) → input features (e.g., Toss, Venue, Outlook)  
- \( C_k \) → output column (Result)  
- \( c_i \in C_k \) → possible outputs (e.g., Won, Lost)

---

##### Example (Your Dataset)

- Features (X): Toss, Venue, Outlook  
- Output (Cₖ): Result  

$$
C_k = \{\text{Won}, \text{Lost}\}
$$

##### Naive Bayes Formula

$$
P(C_k \mid X) = \frac{P(X \mid C_k)\cdot P(C_k)}{P(X)}
$$

---

##### Removing the Denominator

$$
P(C_k \mid X) \propto P(X \mid C_k)\cdot P(C_k)
$$

##### Applying Conditional Probability

From Bayes Theorem:

$$
P(C_k \mid X) \propto P(X \mid C_k)\cdot P(C_k)
$$

---

##### Expand Using Conditional Probability

Let:

$$
X = (x_1, x_2, x_3, \dots, x_n)
$$

Then:

$$
P(C_k \mid X) \propto P(x_1, x_2, x_3, \dots, x_n \mid C_k)\cdot P(C_k)
$$

---

##### Note

Here, we use comma (,) instead of intersection:

$$
(x_1, x_2, x_3) \equiv (x_1 \cap x_2 \cap x_3)
$$

________
_______

##### Expansion Step

$$
P(C_k \mid X) = P(X, C_k)
$$

---

$$
= P(x_1, x_2, x_3, \dots, x_n, C_k)
$$

---

$$
= P(x_1 \mid x_2, x_3, \dots, x_n, C_k)\cdot P(x_2, x_3, \dots, x_n, C_k)
$$

---

Let:

$$
a = P(x_1 \mid x_2, x_3, \dots, x_n, C_k)
$$

---

$$
P(C_k \mid X) = a \cdot P(x_2, x_3, \dots, x_n, C_k)
$$

$$
P(x_2 \mid x_3, \dots, x_n, C_k)
$$

---

$$
= a \cdot P(x_2 \mid x_3, x_4, \dots, x_n, C_k)\cdot P(x_3, x_4, \dots, x_n, C_k)
$$

---

Let:

$$
b = P(x_2 \mid x_3, x_4, \dots, x_n, C_k)
$$

---

$$
= ab \cdot P(x_3, x_4, \dots, x_n, C_k)
$$

##### Chain Rule of Conditional Probability

$$
P(x_1, x_2, x_3, \dots, x_n, C_k)
$$

---

Applying chain rule:

$$
= P(x_1 \mid x_2, x_3, \dots, x_n, C_k)\cdot P(x_2 \mid x_3, \dots, x_n, C_k)\cdots P(x_n \mid C_k)\cdot P(C_k)
$$

---

##### Final Expression

$$
P(C_k \mid X) \propto P(x_1 \mid x_2, x_3, \dots, x_n, C_k)\cdot P(x_2 \mid x_3, \dots, x_n, C_k)\cdots P(x_n \mid C_k)\cdot P(C_k)
$$

## Naive Assumption (Conditional Independence)

In each term, we assume that features are conditionally independent given the class.

---

##### Simplification

$$
P(x_1 \mid x_2, x_3, \dots, x_n, C_k) = P(x_1 \mid C_k)
$$

$$
P(x_2 \mid x_3, x_4, \dots, x_n, C_k) = P(x_2 \mid C_k)
$$

$$
\dots
$$

$$
P(x_n \mid C_k) = P(x_n \mid C_k)
$$

---

##### Final Simplified Form

$$
P(x_1, x_2, x_3, \dots, x_n \mid C_k)
\propto P(x_1 \mid C_k)\cdot P(x_2 \mid C_k)\cdots P(x_n \mid C_k)
$$

> for more simplification we can also write 
$$
P(C_k \mid X) \propto P(C_k)\prod_{i=1}^{n} P(x_i \mid C_k)
$$

___
____
____

##### Naive Bayes with Normalization

$$
P(C_k \mid X) = \frac{1}{Z}\, P(C_k)\prod_{i=1}^{n} P(x_i \mid C_k)
$$

---

##### Maximum A Posteriori (MAP) Rule

$$
\hat{y} = \arg\max_{k \in \{1,2,\dots,K\}} \; P(C_k)\prod_{i=1}^{n} P(x_i \mid C_k)
$$

---

##### What is Z?

$$
Z = P(X)
$$

- It is a normalization constant  
- Ensures probabilities sum to 1  
- Same for all classes → ignored in argmax  

---

##### MAP Intuition

MAP (Maximum A Posteriori) means:

Choose the class that has the highest posterior probability given the data.

---

##### Key Idea

Instead of computing exact probabilities, we compute:

$$
\arg\max \; P(C_k \mid X)
$$

Which simplifies to:

$$
\arg\max \; P(C_k)\prod_{i=1}^{n} P(x_i \mid C_k)
$$

---

##### Conclusion

- MAP selects the most probable class  
- Denominator is ignored because it is constant  
- This is how Naive Bayes makes predictions

##### Problem

Given:

- Outlook = Sunny  
- Temperature = Hot  
- Humidity = High  
- Wind = Weak  

Predict:

$$
\text{PlayTennis = Yes or No}
$$

---

##### Step 1: Prior Probabilities

From dataset:

- Yes = 9  
- No = 5  

$$
P(Yes) = \frac{9}{14}, \quad P(No) = \frac{5}{14}
$$

---

##### Step 2: Likelihoods

###### For Yes (Play = Yes):

- P(Sunny | Yes) = 2/9  
- P(Hot | Yes) = 2/9  
- P(High | Yes) = 3/9  
- P(Weak | Yes) = 6/9  

$$
P(X \mid Yes) = \frac{2}{9} \cdot \frac{2}{9} \cdot \frac{3}{9} \cdot \frac{6}{9}
$$

###### For No (Play = No):

- P(Sunny | No) = 3/5  
- P(Hot | No) = 2/5  
- P(High | No) = 4/5  
- P(Weak | No) = 2/5  

$$
P(X \mid No) = \frac{3}{5} \cdot \frac{2}{5} \cdot \frac{4}{5} \cdot \frac{2}{5}
$$

---

##### Step 3: Apply Naive Bayes

###### For Yes:

$$
P(Yes \mid X) \propto \frac{9}{14} \cdot \left(\frac{2}{9} \cdot \frac{2}{9} \cdot \frac{3}{9} \cdot \frac{6}{9}\right)
$$

$$
= \frac{9}{14} \cdot \frac{72}{6561} \approx 0.0053
$$

---

###### For No:

$$
P(No \mid X) \propto \frac{5}{14} \cdot \left(\frac{3}{5} \cdot \frac{2}{5} \cdot \frac{4}{5} \cdot \frac{2}{5}\right)
$$

$$
= \frac{5}{14} \cdot \frac{48}{625} \approx 0.0274
$$

##### Step 4: Compare

$$
P(Yes \mid X) \approx 0.0053
$$

$$
P(No \mid X) \approx 0.0274
$$

---

##### Final Prediction

$$
\text{PlayTennis = No}
$$
