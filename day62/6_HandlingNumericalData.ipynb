##### Naive Bayes with Numerical Data

For numerical features, we cannot use frequency counts.

Instead, we assume that the data follows a Gaussian (Normal) distribution.

---

##### Gaussian Probability Density Function

$$
P(x \mid C_k) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
$$

---

##### Parameters

- $$\mu$$ → mean of feature for class \( C_k \)  
- $$\sigma^2$$ → variance of feature for class \( C_k \)  

---

##### Final Naive Bayes Formula

$$
P(C_k \mid X) \propto P(C_k)\cdot \prod_{i=1}^{n} P(x_i \mid C_k)
$$

where:

$$
P(x_i \mid C_k) \text{ is computed using Gaussian distribution}$$

##### Gaussian Assumption in Naive Bayes

In Gaussian Naive Bayes, we assume that each feature is normally distributed within each class.

---

##### Formally

For a feature \( x \) and class \( C_k \):

$$
x \mid C_k \sim \mathcal{N}(\mu_k, \sigma_k^2)
$$

---

##### What this means

- Height for Males → follows one Gaussian distribution  
- Height for Females → follows another Gaussian distribution  

---

##### Example

- Male height → mean ≈ 176, variance ≈ some value  
- Female height → mean ≈ 154, variance ≈ some value  

Each class has its own distribution.

---

##### Key Insight

We are not assuming:

- All heights are Gaussian 

We are assuming:

- Heights **within each class** are Gaussian

## using sklearn

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
data = {
    'Height': [172,180,165,190,139,145,160,172],
    'Weight': [150,170,140,200,100,120,140,150],
    'Gender': ['M','M','M','M','F','F','F','F']
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])  # M=1, F=0
df
```

```python
X = df[['Height', 'Weight']]
y = df['Gender']
```

```python
model = GaussianNB()
model.fit(X, y)
```

```python
# Input: Height=185, Weight=170
test = pd.DataFrame([[185,170]], columns=['Height','Weight'])

prediction = model.predict(test)
print("Encoded Prediction:", prediction)
```

```python
print("Predicted Gender:", le.inverse_transform(prediction))
```

##### Is the Gaussian Assumption Always Correct?

No, the assumption that features follow a Gaussian distribution may not always be true.

---

##### Why can it be wrong?

- Real-world data may be:
  - Skewed (not symmetric)
  - Multi-modal (multiple peaks)
  - Contain outliers
  - Not normally distributed

---

##### Example

Height data may look roughly normal, but:

- Income data → highly skewed  
- Age distribution → not always Gaussian  

---

##### Then why does Naive Bayes still work?

Even with incorrect assumptions:

- It estimates probabilities reasonably well  
- Decision boundaries are often still correct  
- Errors may cancel out across features  

---

##### Key Insight

Naive Bayes is a **good classifier**, even if it is a **bad probability estimator**.

---

##### Conclusion

- Assumption may be wrong 
- Model can still perform well 

##### Checking Gaussian Assumption using Distribution Plots

Before applying Gaussian Naive Bayes, we can visualize the data distribution.

---

##### Why?

Gaussian Naive Bayes assumes:

$$
x \mid C_k \sim \mathcal{N}(\mu, \sigma^2)
$$

So we check if the feature approximately follows a normal distribution.

---

##### How to check?

Use distribution plots (distplot / histplot / kdeplot):

- Histogram → shape of data  
- KDE (Kernel Density) → smooth curve  

---

##### Goal

- Bell-shaped curve → suitable for GaussianNB 
- Skewed / irregular → assumption may be weak
