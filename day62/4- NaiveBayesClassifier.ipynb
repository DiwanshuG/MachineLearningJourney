##### Naive Bayes Example

Input:

X = {Toss = Lost, Venue = Mumbai, Outlook = Sunny}

We need to find:

$$
P(\text{Won} \mid X) \quad \text{and} \quad P(\text{Lost} \mid X)
$$

---

##### Step 1: Count from Dataset

Total = 8

- Won = 5  
- Lost = 3  

$$
P(\text{Won}) = \frac{5}{8}, \quad P(\text{Lost}) = \frac{3}{8}
$$

---

##### Step 2: Likelihoods

###### For Won:

From Won rows:

- Toss = Lost → 1/5  
- Venue = Mumbai → 2/5  
- Outlook = Sunny → 3/5  

$$
P(X \mid \text{Won}) = \frac{1}{5} \times \frac{2}{5} \times \frac{3}{5}
$$

---

###### For Lost:

From Lost rows:

- Toss = Lost → 1/3  
- Venue = Mumbai → 1/3  
- Outlook = Sunny → 1/3  

$$
P(X \mid \text{Lost}) = \frac{1}{3} \times \frac{1}{3} \times \frac{1}{3}
$$

---

##### Step 3: Apply Naive Bayes

###### For Won:

$$
P(\text{Won} \mid X) \propto \frac{1}{5} \times \frac{2}{5} \times \frac{3}{5} \times \frac{5}{8}
$$

$$
= \frac{6}{125} \times \frac{5}{8} = \frac{30}{1000} = 0.03
$$

---

###### For Lost:

$$
P(\text{Lost} \mid X) \propto \frac{1}{3} \times \frac{1}{3} \times \frac{1}{3} \times \frac{3}{8}
$$

$$
= \frac{1}{27} \times \frac{3}{8} = \frac{3}{216} \approx 0.0139
$$

---

##### Step 4: Compare

$$
P(\text{Won} \mid X) = 0.03
$$

$$
P(\text{Lost} \mid X) \approx 0.0139
$$

---

##### Final Prediction

$$
\text{Won}
$$

##### Why We Ignore the Denominator in Naive Bayes

Bayes Theorem:

$$
P(Y \mid X) = \frac{P(X \mid Y)\cdot P(Y)}{P(X)}
$$

---

##### What is P(X)?

$$
P(X) = \text{Total probability of observing the input}
$$

Example:

$$
P(X) = P(X \mid \text{Won})P(\text{Won}) + P(X \mid \text{Lost})P(\text{Lost})
$$

---

#####
When we compare classes:

$$
P(\text{Won} \mid X) \quad \text{vs} \quad P(\text{Lost} \mid X)
$$

Both have the same denominator:

$$
P(X)
$$

---

##### So we compare only numerator

Instead of:

$$
\frac{P(X \mid \text{Won})P(\text{Won})}{P(X)} 
\quad vs \quad
\frac{P(X \mid \text{Lost})P(\text{Lost})}{P(X)}
$$

We compare:

$$
P(X \mid \text{Won})P(\text{Won}) 
\quad vs \quad
P(X \mid \text{Lost})P(\text{Lost})
$$

---

##### Final Idea

$$
P(Y \mid X) \propto P(X \mid Y)\cdot P(Y)
$$

(Proportional to)

---

##### Conclusion

We ignore $$P(X)$$ because:

- It is same for all classes  
- It does not change which probability is larger  
- It saves computation  

This is why Naive Bayes is fast and efficient.

##### Why is Naive Bayes called "Naive"?

Naive Bayes is called "naive" because it assumes that all features are independent of each other, given the class label.

---

##### Independence Assumption

$$
P(X_1, X_2, ..., X_n \mid Y) = P(X_1 \mid Y) \cdot P(X_2 \mid Y) \cdots P(X_n \mid Y)
$$

---

##### Why is this Naive?

In real-world data:

- Features are often related
- There can be dependencies between variables

Example:

- "Humidity" and "Rain" are not independent  
- "Venue" and "Weather" may be related  

But Naive Bayes ignores these relationships.

---

##### Why still use it?

Even with this unrealistic assumption:

- It is simple and fast  
- Works well in many real-world problems  
- Requires less data  
- Performs surprisingly well (especially in text classification)

---

##### 

It is called "naive" because it simplifies the problem by assuming independence between features, even when this assumption is not true.

##### Naive Bayes Expansion

We want to find:

$$
P(\text{Won} \mid \text{Lost, Mumbai, Sunny})
$$

---

##### Apply Bayes Theorem

$$
P(\text{Won} \mid X) = \frac{P(X \mid \text{Won}) \cdot P(\text{Won})}{P(X)}
$$

where:

$$
X = (\text{Lost, Mumbai, Sunny})
$$

---

##### Expand Likelihood

$$
P(\text{Lost, Mumbai, Sunny} \mid \text{Won})
$$

Using Naive assumption:

$$
P(\text{Lost} \mid \text{Won}) \cdot P(\text{Mumbai} \mid \text{Won}) \cdot P(\text{Sunny} \mid \text{Won})
$$

---

##### Final Expression

$$
P(\text{Won} \mid \text{Lost, Mumbai, Sunny}) \propto 
P(\text{Lost} \mid \text{Won}) \cdot 
P(\text{Mumbai} \mid \text{Won}) \cdot 
P(\text{Sunny} \mid \text{Won}) \cdot 
P(\text{Won})
$$

---

##### Similarly for Lost

$$
P(\text{Lost} \mid \text{Lost, Mumbai, Sunny}) \propto 
P(\text{Lost} \mid \text{Lost}) \cdot 
P(\text{Mumbai} \mid \text{Lost}) \cdot 
P(\text{Sunny} \mid \text{Lost}) \cdot 
P(\text{Lost})
$$

---

##### Key Idea

We break joint probability into product of individual probabilities using independence:

$$
P(X_1, X_2, X_3 \mid Y) = P(X_1 \mid Y)\cdot P(X_2 \mid Y)\cdot P(X_3 \mid Y)
$$
