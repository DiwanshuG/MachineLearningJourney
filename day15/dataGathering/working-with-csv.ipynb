#### `skiprows` and `nrows` Parameters in `pandas.read_csv()`

#### Overview

The `skiprows` and `nrows` parameters in `pandas.read_csv()` control **which rows** and **how many rows** are read from a CSV (or similar delimited) file. These parameters are especially useful when working with **large datasets**, **sampling data**, or **files containing metadata rows**.

---

#### `skiprows`

#### Description

Specifies rows to skip when reading the file.

#### Syntax

```python
pd.read_csv(filepath, skiprows=...)
```

#### Accepted Values

* **int** – Skip the first *n* rows
* **list of int** – Skip specific row indices
* **callable** – Function that returns `True` for rows to skip

#### Examples

```python
# Skip first 5 rows
pd.read_csv("data.csv", skiprows=5)

# Skip specific rows
pd.read_csv("data.csv", skiprows=[0, 2, 4])

# Skip rows conditionally
pd.read_csv("data.csv", skiprows=lambda x: x % 2 == 0)
```

#### Notes

* Row indexing starts from **0**
* Skipping the header row may require setting `header=None`

---

#### `nrows`

### Description

Limits the number of rows read from the file.

### Syntax

```python
pd.read_csv(filepath, nrows=...)
```

### Accepted Values

* **int** – Number of rows to read

### Examples

```python
# Read first 10 rows only
pd.read_csv("data.csv", nrows=10)
```

---

## Using `skiprows` and `nrows` Together

### Description

Allows reading a **specific range of rows**, useful for batch processing or partial loading.

### Example

```python
# Read rows 100 to 199
pd.read_csv("data.csv", skiprows=100, nrows=100)
```

---

## Use Cases

* Reading large files efficiently
* Ignoring metadata or comments at the top of files
* Previewing datasets
* Processing data in chunks

---

## Summary

| Parameter  | Purpose                    |
| ---------- | -------------------------- |
| `skiprows` | Skips unwanted rows        |
| `nrows`    | Limits number of rows read |

Both parameters help improve **performance**, **memory usage**, and **data handling flexibility**.

````python
import os
import pandas as pd

base_path = "character-encoding-examples"

# Encoding map for each file
encoding_map = {
    "die_ISO-8859-1.txt": "latin-1",
    "portugal_ISO-8859-1.txt": "latin-1",
    "harpers_ASCII.txt": "ascii",
    "olaf_Windows-1251.txt": "cp1251",
    "shisei_UTF-8.txt": "utf-8",
    "yan_BIG-5.txt": "big5",
}

dataframes = {}

for file, enc in encoding_map.items():
    file_path = os.path.join(base_path, file)
    
    df = pd.read_csv(
        file_path,
        encoding=enc,
        sep="\t",                # safe default for txt files
        engine="python",         # tolerant parser
        on_bad_lines="skip"      # skip malformed rows
    )
    
    dataframes[file] = df
    print(f"Loaded {file}: {df.shape}")
````

### 14 . Convertors

````python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return name
````

````python
rename('Royal Challengers Bangalore')
````

````python
pd.read_csv('matches.csv' , converters={'team1':rename})
````
