### Z-Score for Outlier Detection

|Z| > 3  → potential outlier
1) **Assumes** data is normally distributed
2) **Not** reliable for skewed data

### Trimming (Outlier Treatment)

Trimming means removing extreme values (outliers) from a dataset before analysis or modeling.

### Capping (Winsorization)

You define lower and upper limits, and:
1) Values below lower cap → set to lower cap
2) Values above upper cap → set to upper cap
