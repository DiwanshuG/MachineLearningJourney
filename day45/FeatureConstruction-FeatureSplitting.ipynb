# Feature Construction

## What is Feature Construction?
Feature Construction is the process of **creating new features** from existing raw data to make machine learning models **learn patterns more effectively**.

Instead of using data as-is, we **transform, combine, or decompose features** to expose meaningful information that improves model performance.

---

## Why Feature Construction is Important
- Enhances **model accuracy**
- Helps capture **hidden relationships**
- Improves **interpretability**
- Reduces reliance on complex models

> Better features often matter more than better algorithms.

---

## Common Feature Construction Techniques
- **Feature Splitting** ‚Äì breaking one feature into multiple parts  
- **Feature Transformation** ‚Äì log, scaling, normalization  
- **Feature Encoding** ‚Äì label encoding, one-hot encoding  
- **Feature Interaction** ‚Äì combining two or more features  
- **Aggregation** ‚Äì mean, sum, count over groups  

---

## Example
- Date ‚Üí year, month, weekday  
- Full name ‚Üí first name, last name  
- Price ‚Üí log(price)  
- Height & Weight ‚Üí BMI  

---

## Takeaway
Feature Construction bridges the gap between **raw data** and **model intelligence**.  
Well-constructed features can outperform even the most advanced models.


## What is Tidy Data?
Tidy Data is a standard way of organizing datasets so they are **easy to clean, analyze, visualize, and use in machine learning**.

The concept was introduced by **:contentReference[oaicite:0]{index=0}**.

---

## Principles of Tidy Data
A dataset is considered *tidy* if it follows these three rules:

1. Each **variable** forms a **column**
2. Each **observation** forms a **row**
3. Each type of **observational unit** forms a **separate table**

---

## Messy Data Example
| Name | Math | Science |
|------|------|---------|
| A    | 80   | 90      |
| B    | 85   | 88      |

---

## Tidy Data Version
| Name | Subject | Score |
|------|---------|-------|
| A    | Math    | 80    |
| A    | Science | 90    |
| B    | Math    | 85    |
| B    | Science | 88    |

---

## Why Tidy Data is Important
- Simplifies **data cleaning**
- Improves **EDA and visualization**
- Makes **feature engineering** easier
- Works smoothly with **ML pipelines**

---

## Takeaway
Well-structured data leads to **better insights and better models**.  
Tidy data is the foundation of effective data science.

## Applying feature construction

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1
```

```python
X.head()
```

```python
def family_category(num: int) -> int:
    if num == 1:
        return 0      # alone
    if num <= 4:
        return 1      # small family
    return 2          # large family
```

```python
family_category(6)
```

```python
X['Family_type'] = X['Family_size'].apply(family_category)
```

```python
X.head()
```

```python
X = X.drop(columns=['SibSp', 'Parch', 'Family_size'])
```

```python
X.head()
```

```python
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression(max_iter=1000))
])
```

```python
np.mean(cross_val_score(pipe, X, y, cv=5, scoring='accuracy'))
```

```python
X = pd.concat([X, df[['Is_Married', 'Title']]], axis=1)
```

```python
X.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
numeric_features = ['Pclass', 'Age', 'Family_type', 'Is_Married']
categorical_features = ['Title']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ]), numeric_features),

        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ]
)
```

```python
pipe = Pipeline([
    ('preprocess', preprocessor),
    ('model', LogisticRegression(max_iter=1000))
])
```

```python
cv_scores = cross_val_score(
    pipe,
    X,
    y,
    cv=5,
    scoring='accuracy'
)
```

```python
print("CV scores:", cv_scores)
print("Mean CV accuracy:", cv_scores.mean())
print("Std deviation   :", cv_scores.std())
```
