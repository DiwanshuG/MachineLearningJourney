# Ridge Regression (L2 Regularization) – Quick Revision

## What is Ridge Regression?

Ridge Regression is a regularized version of Linear Regression that adds a penalty term to prevent overfitting.

---

## Cost Function

In Linear Regression:

$$
Loss = MSE
$$

In Ridge Regression:

$$
Loss = MSE + \lambda \sum_{i=1}^{n} w_i^2
$$

- $\lambda$ → Regularization parameter (controls penalty strength)  
- $\sum_{i=1}^{n} w_i^2$ → L2 penalty (squared weights)  

---

## Why "L2"?

Because it uses the **L2 norm** (Euclidean norm) of the weight vector:

$$
\|w\|_2^2 = \sum_{i=1}^{n} w_i^2
$$

---

## What Does Ridge Do?

- Shrinks large coefficients  
- Reduces model complexity  
- Controls overfitting  
- Improves generalization  
- Does **not** make coefficients exactly zero  

---

## Effect of Lambda ($\lambda$)

- $\lambda = 0$ → Normal Linear Regression  
- Small $\lambda$ → Slight shrinkage  
- Large $\lambda$ → Strong shrinkage (risk of underfitting)  

---

## When to Use Ridge?

- Multicollinearity (correlated features)  
- Many features  
- Want to reduce variance but keep all features  

---

## Key Idea

Ridge balances:

- Fit to data 
- Simplicity of model  

**Goal → Better Bias–Variance Tradeoff**

# Ridge Regression (Simple Linear Regression) – Derivation

## Model

We start with simple linear regression:

$$
\hat{y}_i = m x_i + b
$$

---

## Ridge Loss Function

$$
L = \sum_{i=1}^{n} (y_i - m x_i - b)^2 + \lambda m^2
$$

Where:
- $m$ = slope  
- $b$ = intercept  
- $\lambda$ = regularization parameter  

---

## Step 1: Partial Derivative w.r.t. b

$$
\frac{\partial L}{\partial b}
=
\frac{\partial}{\partial b}
\left(
\sum_{i=1}^{n} (y_i - m x_i - b)^2
\right)
$$

Derivative:

$$
\frac{\partial L}{\partial b}
=
-2 \sum_{i=1}^{n} (y_i - m x_i - b)
$$

Set equal to zero:

$$
\sum_{i=1}^{n} (y_i - m x_i - b) = 0
$$

---

## Step 2: Partial Derivative w.r.t. m

$$
\frac{\partial L}{\partial m}
=
\frac{\partial}{\partial m}
\left(
\sum_{i=1}^{n} (y_i - m x_i - b)^2 + \lambda m^2
\right)
$$

Derivative:

$$
\frac{\partial L}{\partial m}
=
-2 \sum_{i=1}^{n} x_i (y_i - m x_i - b)
+ 2 \lambda m
$$

Set equal to zero:

$$
\sum_{i=1}^{n} x_i (y_i - m x_i - b)
=
\lambda m
$$

---

## Final System of Equations

$$
\frac{\partial L}{\partial b} = 0
$$

$$
\frac{\partial L}{\partial m} = 0
$$

Solving these gives the Ridge Regression estimates for $m$ and $b$.

---

## Key Insight

The extra term:

$$
\lambda m^2
$$

adds

$$
2\lambda m
$$

to the derivative, which shrinks the slope $m$ and prevents it from becoming too large.


# or Substituting b

From:

$$
\frac{\partial L}{\partial b} = 0
$$

We get:

$$
b = \bar{y} - m \bar{x}
$$

Where:
- $\bar{x}$ = mean of $x$
- $\bar{y}$ = mean of $y$

---

## Substitute b into the Loss Function

Original Ridge Loss:

$$
L = \sum_{i=1}^{n} (y_i - m x_i - b)^2 + \lambda m^2
$$

Substitute $b = \bar{y} - m \bar{x}$:

$$
L = \sum_{i=1}^{n}
\left(
y_i - m x_i - (\bar{y} - m \bar{x})
\right)^2
+ \lambda m^2
$$

---

## Rearranging

$$
L =
\sum_{i=1}^{n}
\left(
(y_i - \bar{y})
- m (x_i - \bar{x})
\right)^2
+ \lambda m^2
$$

---

## Key Insight

After substitution:

- The equation is now written in **centered form**
- Data is expressed relative to their means
- Ridge still adds the penalty term:

$$
\lambda m^2
$$

This makes the slope $m$ shrink toward zero.


## After substituting:

$$
b = \bar{y} - m\bar{x}
$$

The loss becomes:

$$
L =
\sum_{i=1}^{n}
\left[
(y_i - \bar{y}) - m(x_i - \bar{x})
\right]^2
+ \lambda m^2
$$

---

## Take derivative w.r.t. m

$$
\frac{\partial L}{\partial m}
=
-2 \sum_{i=1}^{n}
\left[
(y_i - \bar{y}) - m(x_i - \bar{x})
\right](x_i - \bar{x})
+ 2\lambda m
$$

Set equal to zero:

$$
\lambda m
-
\sum_{i=1}^{n}
\left[
(y_i - \bar{y}) - m(x_i - \bar{x})
\right](x_i - \bar{x})
= 0
$$

---

## Expand the Summation

$$
\lambda m
-
\sum_{i=1}^{n}
(y_i - \bar{y})(x_i - \bar{x})
+
m \sum_{i=1}^{n}
(x_i - \bar{x})^2
= 0
$$

---

## Rearranging Terms

$$
m
\left(
\sum_{i=1}^{n}
(x_i - \bar{x})^2
+
\lambda
\right)
=
\sum_{i=1}^{n}
(y_i - \bar{y})(x_i - \bar{x})
$$

---

## Final Ridge Slope Formula

$$
m
=
\frac{
\sum_{i=1}^{n}
(y_i - \bar{y})(x_i - \bar{x})
}{
\sum_{i=1}^{n}
(x_i - \bar{x})^2
+
\lambda
}
$$


## Effect of Lambda (λ) on the Slope in Ridge Regression

From the Ridge Regression slope formula:

$$
m =
\frac{
\sum_{i=1}^{n}(y_i - \bar{y})(x_i - \bar{x})
}{
\sum_{i=1}^{n}(x_i - \bar{x})^2 + \lambda
}
$$

### Observation

- The **numerator remains constant**.
- The **denominator increases as $\lambda$ increases**.

### What Happens When $\lambda$ Increases?

- Denominator ↑  
- Fraction value ↓  
- Slope $m$ ↓  

Therefore:

> Increasing $\lambda$ reduces the magnitude of the slope.

---

### Extreme Case

- If $\lambda = 0$ → Normal Linear Regression  
- If $\lambda \to \infty$ → $m \to 0$ (line becomes almost flat)

## creating own ridge class

````python
import numpy as np

class RidgebyD:
    
    def __init__(self, alpha=0.1):
        self.alpha = alpha
        self.m = None
        self.b = None
        
    def fit(self, X_train, y_train):
        
        X_mean = np.mean(X_train)
        y_mean = np.mean(y_train)
        
        num = 0
        den = 0
        
        for i in range(len(X_train)):
            num += (y_train[i] - y_mean) * (X_train[i] - X_mean)
            den += (X_train[i] - X_mean) ** 2
        
        # Ridge formula
        self.m = num / (den + self.alpha)
        self.b = y_mean - self.m * X_mean
        
        print("Slope (m):", self.m)
        print("Intercept (b):", self.b)
    
    def predict(self, X_test):
        return self.m * X_test + self.b
````

````python
reg = RidgebyD(alpha=100)
````

````python
reg.fit(X,y)
````

# Ridge Regression for n-dimensions

___________________

For multiple features:

$$
\mathbf{y} = X \mathbf{w}
$$

### Where:

- $X$ → $(n \times p)$ feature matrix  
- $\mathbf{w}$ → $(p \times 1)$ weight vector  
- $\mathbf{y}$ → $(n \times 1)$ target vector  

### Meaning:

- $n$ = number of training samples  
- $p$ = number of features  


### Squared Error in Matrix Form

The squared error term can be written as:

$$
(Xw - y)^T (Xw - y)
$$

### Explanation

- $X$ → $(n \times p)$ feature matrix  
- $w$ → $(p \times 1)$ weight vector  
- $y$ → $(n \times 1)$ target vector  

- $(Xw - y)$ → residual (error) vector  
- $(Xw - y)^T (Xw - y)$ → sum of squared errors  

This is the compact matrix form of:

$$
\sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$


# Ridge Regression – Loss Function (Matrix Form)

## Loss Function

In Ridge Regression, we add an L2 penalty to the normal squared error:

$$
L(w) = (Xw - y)^T (Xw - y) + \lambda w^T w
$$

---

## Explanation

- $X$ → $(n \times p)$ feature matrix  
- $w$ → $(p \times 1)$ weight vector  
- $y$ → $(n \times 1)$ target vector  
- $\lambda$ → regularization parameter  

---

## Terms Breakdown

### 1) Squared Error Term

$$
(Xw - y)^T (Xw - y)
$$

This represents the **Sum of Squared Errors (SSE)**.

---

### 2) Regularization Term

$$
\lambda w^T w
$$

Since:

$$
w^T w = \sum_{j=1}^{p} w_j^2
$$

This is the **L2 penalty**.

---

## Key Idea

Ridge Regression minimizes:

- Prediction error  
- + Magnitude of weights  

Goal → Reduce overfitting by shrinking coefficients.

### 2)  L2 Regularization Term

$$
\lambda w^T w
$$

$$

$$

This penalizes large weights and helps reduce overfitting.


## Expanded Ridge Loss Function

Starting from:

$$
L = (Xw - y)^T (Xw - y) + \lambda w^T w
$$

After expansion:

$$
L =
w^T X^T X w
- 2 w^T X^T y
+ y^T y
+ \lambda w^T w
$$


# Common Scalar vs Vector Derivatives

These derivatives are often memorized because they resemble familiar scalar derivatives.  
However, matrix derivatives do not always behave exactly like scalar ones.

In the following:

- $b$ is a constant scalar  
- $B$ is a constant matrix  

---

## Scalar Derivatives

| Function | Derivative |
|-----------|------------|
| $bx$ | $b$ |
| $bx$ (w.r.t. $b$) | $x$ |
| $x^2$ | $2x$ |
| $b x^2$ | $2bx$ |

---

## Vector / Matrix Derivatives

| Function | Derivative |
|-----------|------------|
| $x^T B$ | $B$ |
| $x^T b$ | $b$ |
| $x^T x$ | $2x$ |
| $x^T B x$ | $2Bx$ |

---

## Reference

For a more comprehensive reference, see:

https://www.math.uwaterloo.ca/~hwolkowi/matrixcookbook.pdf


## Gradient of Ridge Loss Function

Taking derivative of the loss function with respect to $w$:

$$
\frac{\partial L}{\partial w}
=
2X^T X w
-
2X^T y
+
0
+
2\lambda w
=
0
$$

---

### Explanation

- $2X^T X w$ → derivative of $w^T X^T X w$
- $-2X^T y$ → derivative of $-2 w^T X^T y$
- $0$ → derivative of $y^T y$ (constant term)
- $2\lambda w$ → derivative of $\lambda w^T w$

Setting the gradient equal to zero gives the stationary condition.


## Normal Equation for Ridge Regression

After setting the gradient equal to zero and rearranging, we obtain:

$$
(X^T X + \lambda I) w = X^T y
$$

---

### Explanation

- $X^T X$ → Feature covariance matrix  
- $\lambda I$ → Regularization term (identity matrix scaled by $\lambda$)  
- $w$ → Weight vector  
- $X^T y$ → Correlation between features and target  

This is the Ridge Regression normal equation.

### making our own class

````python
class MeraRidgeND:
    
    def __init__(self, alpha=1.0):
        self.alpha = alpha
        self.coef_ = None
        self.intercept_ = None
        
    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        # Add bias column
        X_bias = np.c_[np.ones(n_samples), X]
        
        # Identity matrix
        I = np.eye(n_features + 1)
        I[0, 0] = 0  # do not regularize intercept
        
        # Closed form solution
        theta = np.linalg.inv(
            X_bias.T @ X_bias + self.alpha * I
        ) @ X_bias.T @ y
        
        self.intercept_ = theta[0]
        self.coef_ = theta[1:]
        
    def predict(self, X):
        return X @ self.coef_ + self.intercept_
````

````python
my_model = MeraRidgeND(alpha=1.0)
my_model.fit(X_train, y_train)
````

````python
y_pred_custom = my_model.predict(X_test)

print("Custom R2:", r2_score(y_test, y_pred_custom))
print("Custom RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_custom)))
````
