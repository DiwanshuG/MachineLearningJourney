### MAE : [mean absolute error]

MAE measures the average absolute difference between actual and predicted values. 
$$
MAE = \frac{1}{n}\sum \left| y_{\text{actual}} - y_{\text{predicted}} \right|
$$

1) Advantages of MAE
- Easy to understand and interpret
- Error is in the same unit as the target variable
- Treats all errors equally
- Less sensitive to outliers compared to MSE/RMSE

2) Disadvantages of MAE

- Does not penalize large errors heavily
- Not differentiable at zero (can be harder for some optimization methods)
- May not reflect model performance well when large errors are critical

____

$$

### RMSE

RMSE is the square root of MSE, bringing the error back to the original unit of the target variable.

$$
RMSE = \sqrt{\frac{1}{n}\sum (y_{\text{actual}} - y_{\text{predicted}})^2}
$$



##### Advantages of RMSE
- Easy to interpret since it is in the same unit as the target variable
- Penalizes large errors more than small errors
- Commonly used for comparing regression models
- Reflects overall model performance effectively

##### Disadvantages of RMSE
- Sensitive to outliers
- Can be overly influenced by large errors
- Not as robust as MAE when outliers are present

##### Numerator
$$
\sum (y_{\text{actual}} - y_{\text{predicted}})^2
$$

- This is **SSR (Sum of Squared Residuals)**
- Represents the **error made by the model**
- Lower value indicates better predictions

---

##### Denominator
$$
\sum (y_{\text{actual}} - \bar{y})^2
$$

- This is **SST (Total Sum of Squares)**
- Represents the **total variance in the data**
- Baseline model predicts the **mean** for all values

---

##### Ratio Interpretation
$$
\frac{SSR}{SST}
$$

- Fraction of variance **not explained** by the model

---

##### Final Interpretation
$$
R^2 = 1 - \frac{SSR}{SST}
$$

- Fraction of variance **explained** by the model

---


The closer the RÂ² score is to **1**, the more **perfectly** the model fits the data.
