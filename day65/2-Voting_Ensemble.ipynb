# Voting Ensemble 

### Voting Ensemble – Intuition

Assume we have three models:

m1 → m2 → m3

* These models can be of the **same algorithm** or **different algorithms**
* All models are trained on the **same dataset**

### Training Phase

* The dataset is given to all three models
* Each model learns patterns independently

### Prediction Phase

* A new data point (query) is given to all models
* Each model makes its own prediction

### Final Prediction

#### For Classification:

* Use **majority voting**
* The class predicted by most models is selected

#### For Regression:

* Use **averaging (mean)**
* Final prediction = average of all model predictions

### Insight

Even if individual models are weak or make errors, combining their predictions leads to a more accurate and stable result.

## Why Voting Ensemble Works (Intuition + Proof)

Assume we have 3 independent models:

* m1, m2, m3
* Each model has accuracy = **0.7 (70%)**
* Error rate = **0.3 (30%)**

### Assumptions

* Models are **independent**
* Each model performs **better than random (> 50%)**

---

## Goal

We want to find:

What is the probability that the **ensemble (majority voting)** gives the correct prediction?

---

## Key Idea

For 3 models, the final prediction is correct if:

* At least **2 models are correct**

---

## Possible Correct Cases

### Case 1: All 3 models are correct

P(all correct) =
0.7 × 0.7 × 0.7 = **0.343**

---

### Case 2: Exactly 2 models are correct

There are 3 such combinations:

* m1 correct, m2 correct, m3 wrong
* m1 correct, m2 wrong, m3 correct
* m1 wrong, m2 correct, m3 correct

Each case probability:
0.7 × 0.7 × 0.3 = **0.147**

Total for all 3 cases:
3 × 0.147 = **0.441**

---

## Final Accuracy of Voting Ensemble

Total probability of correct prediction:

0.343 + 0.441 = **0.784 (78.4%)**

---

## Conclusion

* Individual model accuracy = **70%**
* Ensemble accuracy = **78.4%**

The ensemble performs **better than any single model**.

---

## Why This Happens

* Models make **different mistakes**
* Majority voting cancels out errors
* Correct predictions dominate when models are reasonably accurate

---

## Key Insight

> If base models are independent and better than random (>50%), combining them using voting increases overall accuracy.

---

## Important Note

If models are:

* Highly correlated → benefit reduces
* Worse than random (<50%) → ensemble may fail

So **diversity + decent accuracy** are essential.
