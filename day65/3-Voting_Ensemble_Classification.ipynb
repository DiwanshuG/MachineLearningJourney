# Voting Classifier: Hard Voting vs Soft Voting

A Voting Classifier combines predictions from multiple models to produce a final output.

By default, the voting method is **hard voting**, but **soft voting** is often more powerful when probability outputs are available.

---

## 1) Hard Voting (Majority Voting)

In hard voting:

* Each model predicts a **class label** (0 or 1, not probabilities)
* The final prediction is based on the **majority class**

### Example

Assume we have 2 models:

* Model 1 predicts:
  Class 0 → 0.6
  Class 1 → 0.4
  → Final label = **0**

* Model 2 predicts:
  Class 0 → 0.8
  Class 1 → 0.2
  → Final label = **0**

### Final Decision (Hard Voting):

* Model 1 → 0
* Model 2 → 0
* **Final Prediction → 0 (majority vote)**

👉 Probabilities are ignored — only final class labels matter.

---

## 2) Soft Voting (Probability-Based Voting)

In soft voting:

* Each model provides **probabilities for each class**
* We take the **average probability** across models
* The class with the highest average probability is selected

### Example

Using the same models:

#### Step 1: Collect Probabilities

| Model   | Class 0 | Class 1 |
| ------- | ------- | ------- |
| Model 1 | 0.6     | 0.4     |
| Model 2 | 0.8     | 0.2     |

#### Step 2: Average Probabilities

* Class 0 → (0.6 + 0.8) / 2 = **0.7**
* Class 1 → (0.4 + 0.2) / 2 = **0.3**

#### Step 3: Final Prediction

* **Class 0** (higher average probability)

---

## Key Difference

| Aspect           | Hard Voting   | Soft Voting                |
| ---------------- | ------------- | -------------------------- |
| Input            | Class labels  | Probabilities              |
| Method           | Majority vote | Average probabilities      |
| Information used | Less          | More (confidence included) |
| Performance      | Good          | Usually better             |

---


> Hard voting ignores confidence, while soft voting leverages probability information, making it more reliable in most cases.


## Hard Voting vs Soft Voting – Practical Insight

It is **not necessary** that soft voting will always perform better than hard voting.

* Soft voting **often performs better** because it uses probability (confidence) information
* However, its performance depends on how well the models estimate probabilities

---

## When Soft Voting May Not Perform Better

* If model probabilities are **poorly calibrated**
* If some models are **overconfident but wrong**
* If models have very different probability distributions

---

## Best Practice (Real-World Approach)

In real-world scenarios:

* Try both **hard voting** and **soft voting**
* Evaluate them using proper metrics (accuracy, F1-score, etc.)
* Choose the one that performs better on validation data

---

> Soft voting is usually better, but the best approach is always to experiment and select the method based on actual performance.
