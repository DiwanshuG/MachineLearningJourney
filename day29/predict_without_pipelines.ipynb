```python
import pickle
import numpy as np
```

```python
# load encoders
ohe_sex = pickle.load(open('models/ohe_sex.pkl', 'rb'))
ohe_embarked = pickle.load(open('models/ohe_embarked.pkl', 'rb'))

# load model
clf = pickle.load(open('models/clf.pkl', 'rb'))
```

```python
# Assume user input
# PClass / Sex / Age / SibSp / Parch / Fare / Embarked
test_input = np.array([2, 'male', 31.0, 0, 0, 10.5, 'S'],dtype=object).reshape(1, 7)
```

```python
test_input
```

```python
test_input_sex = ohe_sex.transform(test_input[:,1].reshape(1,1))
```

```python
test_input_sex
```

```python
# here the warning says : “I don’t see feature names, but I remember them.”
```

```python
test_input_embarked  = ohe_embarked.transform(test_input[:,-1].reshape(1,1))
```

```python
test_input_embarked
```

### “Here, we have to do all of this manually because we’re not using pipelines.”

```python
test_input_age = test_input[:,2].reshape(1,1)
```

```python
test_input_transformed = np.concatenate(
    (
        test_input[:, [0, 3, 4, 5]],  # Pclass, SibSp, Parch, Fare
        test_input_age,               # imputed Age
        test_input_sex,               # encoded Sex
        test_input_embarked            # encoded Embarked
    ),
    axis=1
)
```

```python
test_input_transformed.shape
```

```python
clf.predict(test_input_transformed)
```
