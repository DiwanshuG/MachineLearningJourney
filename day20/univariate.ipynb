```python
import pandas as pd
df = pd.read_csv('train.csv')
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.head()
```

### 1) Categorical Dataset
#### a) Countplot 

```python
# Survived & Pclass 

plt.figure(figsize=(7, 5))

sns.countplot(
    data=df,
    x="Pclass",
    hue="Survived",
    palette="viridis"
)

plt.title("Survival by Passenger Class", fontsize=14, weight="bold")
plt.xlabel("Passenger Class")
plt.ylabel("Count")

sns.despine()
plt.tight_layout()
plt.show()
df['Survived'].value_counts()
```

```python
df['Pclass'].value_counts()
```

```python
df['Sex'].value_counts()
```

```python
counts = df["Embarked"].value_counts().reset_index()
counts.columns = ["Embarked", "Count"]

plt.figure(figsize=(6,4))
sns.barplot(
    data=counts,
    x="Embarked",
    y="Count",
    palette="viridis"
)

plt.title("Passenger Count by Embarkation Port")
plt.xlabel("Embarked")
plt.ylabel("Count")
sns.despine()
plt.tight_layout()
plt.show()
```

### b) Piechart

```python
data = df["Embarked"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    data,
    labels = data.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Passenger Distribution by Embarkation Port")
plt.axis("equal")
plt.show()
```

```python
dataClass = df["Pclass"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    dataClass,
    labels = dataClass.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Passenger Distribution by Class")
plt.axis("equal")
plt.show()
```

```python
survived_df = df[df['Survived']==1]
class_count = survived_df["Pclass"].value_counts().sort_index()
plt.figure(figsize=(6,6))
plt.pie(
    class_count,
    labels=[
        f"Class {cls}" for cls in class_count.index
    ],
    autopct="%1.1f%%",
    startangle=90
)

plt.axis("equal")
plt.title("Survived Passengers by Travel Class (%)", fontsize=14, weight="bold")
plt.show()
```

```python
dataSex = df["Sex"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    dataSex,
    labels = dataSex.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Passenger Distribution by Gender")
plt.axis("equal")
plt.show()
```
