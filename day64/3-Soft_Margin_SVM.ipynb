##### Soft Margin SVM

- In real-world datasets, data is rarely perfectly separable  
- Hard Margin SVM fails when:
  - There is noise  
  - There are outliers  
  - Classes overlap  

- To handle this, we use **Soft Margin SVM**

---

##### 

- Allow some points to:
  - Lie inside the margin  
  - Even be misclassified  

- But **penalize** these violations  

---


- Maximizing a function is equivalent to minimizing its reciprocal:

$$
\max f(x) \;\Longleftrightarrow\; \min \frac{1}{f(x)}
$$

- Applying this idea to SVM margin:

$$
\arg\min_{w,b} \; \frac{\|w\|}{2}
$$

##### Insight

- Instead of maximizing the margin directly,  
  we minimize $\|w\|$ (or $\frac{1}{2}\|w\|^2$ for convenience)  
- This makes the optimization problem easier to solve  

##### Interpretation of Slack Variable $\xi_i$

- The slack variable $\xi_i$ represents the **amount of violation** of the margin constraint for a data point  

##### Intuition

- $\xi_i$ can be understood as the **distance of a point in the wrong direction** from its correct hyperplane  

- Specifically:
  - For a point belonging to the positive class:
    - $\xi_i$ measures how far it is from the **positive hyperplane ($\pi^+$)** if it violates the margin  

##### Cases

- $\xi_i = 0$  
  - Point is correctly classified and lies outside the margin  

- $0 < \xi_i < 1$  
  - Point is inside the margin but still correctly classified  

- $\xi_i \geq 1$  
  - Point is misclassified (on the wrong side of the decision boundary)  

##### Key Insight

- $\xi_i$ quantifies how much a point **breaks the SVM constraints**  
- Larger $\xi_i$ → larger violation → higher penalty in optimization  

##### Introducing Slack Variables

- For each data point, introduce a slack variable $\xi_i$:

$$
y_i (w^T x_i + b) \geq 1 - \xi_i
$$

- Where:
  - $\xi_i = 0$ → correctly classified, outside margin  
  - $0 < \xi_i < 1$ → inside margin but correct side  
  - $\xi_i \geq 1$ → misclassified  

---


##### Why Our Equation Look Different

- According to (what we wrote):

$$
\arg\min_{w,b} \; \frac{\|w\|}{2} + \sum_{i=1}^{n} \xi_i
$$

- Standard form :

$$
\text{Minimize } \frac{1}{2} \|w\|^2 + C \sum_{i=1}^{n} \xi_i
$$

---

##### Key Differences Explained

###### 1) $\|w\|$ vs $\|w\|^2$

- Both aim to **minimize the magnitude of $w$**
- Using $\|w\|^2$ is preferred because:
  - It is **smooth and differentiable**
  - Easier for optimization algorithms  

- Both lead to the **same optimal solution**

---

###### 2) Missing $C$ (Regularization Parameter)

- version we wrote:

$$
\frac{\|w\|}{2} + \sum \xi_i
$$

- Standard version:

$$
\frac{1}{2} \|w\|^2 + C \sum \xi_i
$$

- $C$ controls the trade-off:
  - Margin size vs classification error  

- If $C = 1$, both forms become similar  

---

##### Final Standard Form (Most Used)

$$
\text{Minimize } \frac{1}{2} \|w\|^2 + C \sum_{i=1}^{n} \xi_i
$$

$$
\text{Subject to } \; y_i (w^T x_i + b) \geq 1 - \xi_i
$$

$$
\xi_i \geq 0
$$

---

##### Key Insight

- These are **equivalent formulations** with slight variations  
- The standard form is preferred because:
  - Easier to optimize  
  - More flexible (due to $C$)  


##### Role of C (Regularization Parameter)

- $C$ controls the trade-off between:
  - Maximizing margin  
  - Minimizing classification error  

- High $C$:
  - Penalizes errors heavily  
  - Less tolerance → behaves like Hard Margin  

- Low $C$:
  - Allows more violations  
  - Better generalization

##### Comparison: SVM vs Logistic Regression Objective

- Soft Margin SVM objective:

$$
\text{Minimize } \frac{1}{2} \|w\|^2 + C \sum_{i=1}^{n} \xi_i
$$

- Logistic Regression objective:

$$
\text{Logistic Loss} + \lambda \|w\|^2
$$

---

##### Understanding the Similarity

- Both models have two main components:

###### 1) Loss Term (Data Fit)

- **SVM**:
  - Uses hinge-like loss (via $\xi_i$)
  - Penalizes points that violate the margin  

- **Logistic Regression**:
  - Uses logistic loss
  - Penalizes incorrect predictions smoothly  

---

###### 2) Regularization Term

- Both include:

$$
\|w\|^2
$$

- This term:
  - Controls model complexity  
  - Prevents overfitting  
  - Encourages smaller weights  

---

##### Role of Hyperparameters

- In SVM:
  - $C$ controls penalty on violations  

- In Logistic Regression:
  - $\lambda$ controls strength of regularization  

- Relationship:
  - $C \uparrow$ → less regularization  
  - $\lambda \uparrow$ → more regularization  

---

##### Key Difference

- **SVM**:
  - Focuses on maximizing margin  
  - Only cares about critical points (support vectors)  

- **Logistic Regression**:
  - Focuses on probability estimation  
  - Uses all data points  

---

##### Core Insight

- Both follow the same structure:
  - **Loss + Regularization**  
- But differ in:
  - Type of loss  
  - Interpretation of output
