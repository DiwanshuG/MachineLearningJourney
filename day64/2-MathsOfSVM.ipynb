##### Hyperplane Optimization in SVM

- In this part, we focus on finding the **optimal hyperplane (decision boundary)**  

- The goal of SVM is:
  - To choose a hyperplane such that the **margin (d) is maximum**  

- Margin (d):
  - The distance between the two parallel planes (π⁺ and π⁻)  
  - These planes touch the closest data points (support vectors)  

- The optimal hyperplane is the one that:
  - Maximizes the distance between π⁺ and π⁻  
  - Ensures better generalization on unseen data  

##### 

- Instead of just separating the data,
  SVM finds the hyperplane that separates data with the **maximum possible margin (d)**

##### Intuition Using Vector Projection for Decision Boundary

- Consider a vector $w$ that is **perpendicular (normal)** to the hyperplane  
- This vector points from the origin towards the hyperplane  

- Now, take a new data point $(x, y)$ (or generally $x$ in n-dimensions)  

- To determine which side of the hyperplane this point lies on:
  - Project the point onto the vector $w$  
  - Let this projection be denoted as vector $u$  

##### Key Idea

- The projection of the point onto $w$ gives us a measure of:
  - How far the point lies along the direction of the hyperplane normal  

- If the projection lies:
  - On the same side as the hyperplane → point belongs to the positive class  
  - On the opposite side → point belongs to the negative class  

##### Mathematical Condition

- The dot product between $w$ and $x$ determines the side:

$$
w \cdot x \; \gtrless \; c
$$

- Where:
  - $c$ represents a threshold related to the hyperplane position  

##### Interpretation

- If $w \cdot x > c$ → point lies on the positive side  
- If $w \cdot x < c$ → point lies on the negative side  

##### Core Insight

- SVM uses this projection idea to convert geometry into algebra  
- The decision is ultimately based on the **sign and magnitude of $w \cdot x$**

------
------

##### Converting Geometric Intuition into Equations

- From the projection idea, we compare the dot product with a threshold:

$$
w \cdot u \geq c
$$

- Rearranging the inequality:

$$
w \cdot u - c \geq 0
$$

- Let us replace $-c$ with a bias term $b$:

$$
w \cdot u + b \geq 0
$$

##### Final Decision Function

- This leads to the standard SVM decision rule:

$$
w^T x + b \geq 0
$$

##### Insight

- The term $w^T x + b$ determines:
  - Which side of the hyperplane the point lies on  
  - The sign → class label  
  - The magnitude → distance from the boundary  

##### Example: Determining the Side of a Point

- Given hyperplane equation:

$$
2x + 3y + 3 = 0
$$

- This can be written in standard form:

$$
w^T x + b = 0
$$

- Where:
  - $w = (2, 3)$  
  - $b = 3$  

##### Test a Point

- Let the point be $(3, 2)$  

- Substitute into the decision function:

$$
2(3) + 3(2) + 3
$$

$$
= 6 + 6 + 3 = 15 \geq 0
$$

##### Conclusion

- Since the value is **greater than 0**:
  - The point lies on the **positive side** of the hyperplane  

##### Key Insight

- If $w^T x + b \geq 0$ → Positive side  
- If $w^T x + b < 0$ → Negative side

#####  Decision Rule
- For any data point $x_i$, the prediction is based on:

$$
\hat{y} =
\begin{cases}
+1 & \text{if } w^T x_i + b \geq 0 \\
-1 & \text{if } w^T x_i + b < 0
\end{cases}
$$

##### Idea 

- Compute the value of $w^T x_i + b$  
- The **sign** of this value determines the class:
  - Positive → Class +1  
  - Negative → Class -1  

##### Intuition

- The expression $w^T x_i + b$ tells:
  - Which side of the hyperplane the point lies on  
  - How far the point is from the decision boundary  

- Larger magnitude → more confident classification  

> $w^T x$ is just a **matrix/vector notation** for the dot product  

> $$
  w^T x = w \cdot x
$$

## Assumptions for Hyperplanes in SVM

- The central decision boundary (main hyperplane) is defined as:

$$
w^T x + b = 0
$$

- Two parallel hyperplanes are constructed on either side of the decision boundary:

  - Positive side hyperplane:
  $$
  w^T x + b = +1
  $$

  - Negative side hyperplane:
  $$
  w^T x + b = -1
  $$

##### Interpretation

- These two hyperplanes pass through the **closest data points** (support vectors)  
- The region between them defines the **margin**  

- Distance between:
  - $w^T x + b = +1$  
  - $w^T x + b = -1$  
  is the margin $d$  


##### Why Do We Assume Hyperplanes as $w^T x + b = \pm 1$?

##### 1) Why do we assume equations in this form?

- The general equation of a hyperplane is:

$$
w^T x + b = 0
$$

- This equation is **not unique**:
  - If $(w, b)$ is a solution, then $(2w, 2b)$, $(10w, 10b)$ are also valid  
- So, we **fix a scale** to make the solution unique  

- SVM introduces a constraint:
  - Closest points (support vectors) satisfy:

$$
w^T x + b = +1 \quad \text{and} \quad w^T x + b = -1
$$

- This is called **normalization of the hyperplane**

---

##### 2) Why are magnitudes equal (+1 and -1)?

- Because the decision boundary lies exactly **in the middle**:

$$
w^T x + b = 0
$$

- If we used unequal values (e.g., +1 and -2):
  - The boundary would not be centered  
  - Margin would be asymmetric  

- Using +1 and -1 ensures:
  - Equal distance from the central hyperplane  
  - Symmetric margin on both sides  

---

##### 3) Why specifically ±1 and not ±25 or any other number?

- We **can** use ±25, ±100, etc., but:

- If we scale everything:
  - $25(w^T x + b) = \pm 25$  
  - This is equivalent to:

$$
w^T x + b = \pm 1
$$

- So, larger values just **scale $w$ and $b$**, but do not change the geometry  

- Choosing ±1:
  - Keeps equations simple  
  - Makes margin formula clean  
  - Standard convention in SVM  

---

##### Key Insight

- The choice of ±1 is not special mathematically  
- It is a **normalization trick** to:
  - Remove ambiguity  
  - Simplify optimization  
  - Ensure symmetry of the margin

##### Finding the Margin (d) in SVM

- Now, the goal is to find the equation of the margin $d$  

- Consider two classes of data points:
  - Green points (one class)  
  - Red points (other class)  

- We construct two parallel hyperplanes:
  - $\pi^+$ for the positive class  
  - $\pi^-$ for the negative class  

##### Constraints

- The key constraint is:
  - No data point should lie between $\pi^+$ and $\pi^-$  

- This means:
  - All green points must lie **on or beyond** $\pi^+$  
  - All red points must lie **on or beyond** $\pi^-$  

- In simple terms:
  - Data points cannot cross into the margin region  

##### Objective

- Under these constraints, we aim to:

$$
\text{maximize } d
$$

- Where $d$ is the distance between $\pi^+$ and $\pi^-$  

##### Key Idea

- SVM finds the maximum possible margin  
- While ensuring that:
  - All data points are correctly separated  
  - No point lies inside the margin  

- This leads to a more robust and generalizable model  

----
---

##### Writing Constraints Mathematically in SVM

- We have two hyperplanes:

$$
w^T x + b = +1 \quad (\pi^+)
$$

$$
w^T x + b = -1 \quad (\pi^-)
$$

##### Constraints for Data Points

- For points belonging to the **positive class** ($y_i = +1$):

$$
w^T x_i + b \geq +1
$$

- For points belonging to the **negative class** ($y_i = -1$):

$$
w^T x_i + b \leq -1
$$

##### Combined Constraint

- These can be written in a single equation:

$$
y_i (w^T x_i + b) \geq 1
$$

##### Notes:

- This constraint ensures that:
  - All points lie **outside or on the margin boundaries**  
  - No point lies between $\pi^+$ and $\pi^-$  

- This is the fundamental constraint used in SVM optimization

----

##### Constraint and Support Vectors in SVM

- The constraint for all data points is:

$$
y_i (w^T x_i + b) \geq 1
$$

- For **support vectors** (points lying exactly on the margin):

$$
y_i (w^T x_i + b) = 1
$$

##### Margin (d) Using Support Vectors

- Consider two support vectors:
  - $x_1$ lying on the negative hyperplane $(\pi^-)$  
  - $x_2$ lying on the positive hyperplane $(\pi^+)$  

- Form a vector between them:

$$
x_2 - x_1
$$

##### Key Idea

- This vector connects the two closest points from opposite classes  
- However, the margin is not the full length of this vector  

- The margin $d$ is the **projection of this vector onto the normal vector $w$**  
  - Since $w$ is perpendicular to the hyperplane  

##### Interpretation

- It does not matter where exactly $x_1$ and $x_2$ are located along the hyperplanes  
- Only the component of $(x_2 - x_1)$ in the direction of $w$ matters  

- When we project $(x_2 - x_1)$ onto $w$, we get:

$$
d = \text{projection of } (x_2 - x_1) \text{ onto } w
$$

##### Core Insight

- The margin is essentially the **distance between the two hyperplanes measured along the normal direction**



- The margin $d$ can be computed using the projection of the vector $(x_2 - x_1)$ onto the normal vector $w$:

$$
d = \frac{(x_2 - x_1) \cdot w}{\|w\|}
$$

##### Terms Explanation

- $x_1$ → support vector on $\pi^-$  
- $x_2$ → support vector on $\pi^+$  
- $w$ → normal (perpendicular) vector to the hyperplane  
- $\|w\|$ → magnitude (length) of vector $w$  


##### Derivation of Margin (d) in SVM

- We start with the formula for margin using projection:

$$
d = \frac{(x_2 - x_1) \cdot w}{\|w\|}
$$

##### Expanding the Numerator

$$
d = \frac{x_2 \cdot w - x_1 \cdot w}{\|w\|}
$$

##### Using Support Vector Conditions

- For support vectors:

$$
w^T x_2 + b = +1
$$

$$
w^T x_1 + b = -1
$$

- Rearranging:

$$
w^T x_2 = 1 - b
$$

$$
w^T x_1 = -1 - b
$$

##### Substitute into Margin Equation

$$
d = \frac{(1 - b) - (-1 - b)}{\|w\|}
$$

$$
d = \frac{1 - b + 1 + b}{\|w\|}
$$

$$
d = \frac{2}{\|w\|}
$$

##### Final Result

$$
d = \frac{2}{\|w\|}
$$

##### Key Insight

- The margin depends only on the magnitude of $w$  
- Maximizing the margin is equivalent to minimizing $\|w\|$  
- This leads to the core optimization objective in SVM  

#####  Hard Margin SVM

- Hard Margin SVM is the **basic version of SVM** where:
  - All data points must be **perfectly classified**
  - No errors or misclassifications are allowed  

##### Key Assumption

- The data must be **linearly separable**  
  - There exists a hyperplane that separates the classes without any overlap  

##### Optimization Problem

$$
\text{Minimize } \frac{1}{2} \|w\|^2
$$

$$
\text{Subject to } \; y_i (w^T x_i + b) \geq 1 \quad \forall i
$$

##### Characteristics

- Strict constraint:  
  - Every point lies **outside or on the margin**  
- No tolerance for:
  - Noise  
  - Outliers  

##### Limitations

- Not suitable for real-world data because:
  - Most datasets are **not perfectly separable**  
  - Even a single outlier can break the model  

##### SVM Optimization Problem

- From the margin derivation, we know:

$$
d = \frac{2}{\|w\|}
$$

- To maximize the margin, we need to:

$$
\arg\max_{w, b} \; \frac{2}{\|w\|}
$$

##### Subject to Constraint

$$
y_i (w^T x_i + b) \geq 1 \quad \forall i
$$

##### Why Do We See Different Optimization Forms in Hard Margin SVM?

##### 1) What we originally derived

- From geometry, we got:

$$
d = \frac{2}{\|w\|}
$$

- So naturally, we want to maximize the margin:

$$
\arg\max_{w,b} \; \frac{2}{\|w\|}
$$

---

##### 2) Why this form is not used directly

- The expression $\frac{2}{\|w\|}$ is:
  - Non-linear  
  - Difficult to optimize directly  

- Optimization algorithms prefer:
  - Smooth  
  - Convex  
  - Differentiable functions  

---

##### 3) Key Trick (Transformation)

- Maximizing $\frac{2}{\|w\|}$ is equivalent to:

$$
\min \|w\|
$$

- And for mathematical convenience, we further convert it to:

$$
\min \frac{1}{2} \|w\|^2
$$

- Why square?
  - Removes square root  
  - Easier differentiation  
  - Same minimum point  

---

##### 4) Final Hard Margin SVM Optimization

$$
\text{Minimize } \frac{1}{2} \|w\|^2
$$

$$
\text{Subject to } \; y_i (w^T x_i + b) \geq 1
$$

---

##### 5) Big Picture

- These are **not different problems**, but the **same objective written differently**:

- Geometry view:
  - Maximize margin → $\frac{2}{\|w\|}$  

- Optimization view:
  - Minimize complexity → $\|w\|^2$  

---

##### Key Insight

- SVM converts a geometric idea (maximize distance)  
  into an optimization problem (minimize $\|w\|^2$)  
- This makes it solvable using efficient optimization techniques
