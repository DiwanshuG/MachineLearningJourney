#####  Support Vector Machines (SVM)

##### Why SVM is Highly Respected
- SVM holds a strong reputation in the ML community
- Known for being mathematically elegant and powerful
- Considered one of the most robust classical algorithms

##### Robustness
- Works well even in high-dimensional spaces
- Effective when the number of features > number of samples
- Focuses only on critical data points (support vectors) → ignores noise to some extent

##### Applications
- Used in a wide range of scenarios:
  - Text classification (spam detection)
  - Image classification
  - Bioinformatics (gene classification)
  - Face detection
- Performs well in both linear and non-linear problems

##### Historical Context
- Became popular in the 1990s
- Before deep learning, SVM was one of the go-to algorithms for classification tasks

##### Relation to Logistic Regression
- Can be seen in the same family as Logistic Regression
- Both are:
  - Linear classifiers (in basic form)
  - Used for classification problems
- Key difference:
  - Logistic Regression → focuses on probability estimation
  - SVM → focuses on maximum margin separation
- SVM can be considered an improvement over Logistic Regression in terms of:
  - Better handling of complex boundaries (via kernels)
  - Stronger generalization in many cases

#####  
- SVM doesn’t just classify — it classifies with maximum confidence (margin)

## Geometric Intuition 

##### Support Vector Machines (SVM) Decision Boundary Intuition

- Suppose there are multiple possible lines/planes (e.g., π₁, π₂) that can separate the data  
- SVM selects the best possible boundary among them  
- “Best” means the one that maximizes the margin (distance from closest data points)  

##### Decision Function

$$
\sigma(w^T x_i) = P(y = 1)
$$

- $x_i$ represents a data point  
- $w^T x_i$ represents the distance (in a transformed sense) from the decision boundary  

##### Key Insight About Distance

- The value of $w^T x_i$ is proportional to how far the point is from the hyperplane  
- If the point is far from the boundary:  
  - Model is more confident  
  - Probability moves closer to 1 or 0  
- If the point is close to the boundary:  
  - Model is less confident  
  - Probability moves closer to 0.5  

##### Important Clarification

- The sigmoid function $\sigma(w^T x_i)$ is mainly associated with Logistic Regression  
- In SVM:  
  - We focus on maximizing the margin  
  - Not directly on probability  

##### Core Idea

- SVM chooses the boundary where the minimum distance to nearest points is maximized  
- These nearest points are called support vectors  

### Core Objective of SVM

- We have to find the hyperplane which maximizes the margin [Margin Maximizing Hyperplane]



 ##### Important Terms in SVM

- **Hyperplane**  
  - A decision boundary that separates different classes  
  - In 2D: a line, in 3D: a plane, in higher dimensions: a hyperplane  

- **Margin**  
  - The distance between the hyperplane and the nearest data points from each class  
  - SVM aims to maximize this margin  

- **Support Vectors**  
  - The data points that are closest to the hyperplane  
  - These points directly influence the position of the decision boundary  

- **Decision Boundary**  
  - The final separating line/plane chosen by SVM to classify data  

- **Kernel**  
  - A function used to transform data into a higher dimension  
  - Helps to handle non-linear data  

- **Kernel Trick**  
  - A technique that allows SVM to work in higher dimensions without explicitly computing the transformation  

- **Regularization Parameter (C)**  
  - Controls the trade-off between maximizing margin and minimizing classification error  
  - High C → less margin, fewer misclassifications  
  - Low C → larger margin, allows some misclassifications  

- **Gamma (γ)**  
  - Defines how far the influence of a single data point reaches (used in RBF kernel)  
  - High gamma → complex boundary  
  - Low gamma → smoother boundary  

- **Linear SVM**  
  - Used when data is linearly separable  

- **Non-Linear SVM**  
  - Uses kernels to separate data that is not linearly separable

##### How SVM Finds the Best Hyperplane

- Start with a candidate separating hyperplane that divides the data  

- From this hyperplane, draw two parallel planes (or lines in 2D):
  - Move one plane **towards the positive class**
  - Stop when it touches the **first data point**
  - This plane is called **π⁺**

- Similarly:
  - Move another parallel plane **towards the negative class**
  - Stop when it touches the **first data point**
  - This plane is called **π⁻**

- These two planes (π⁺ and π⁻) are the closest boundaries that still correctly separate the data  

- The distance between π⁺ and π⁻ is called the **margin**

- SVM tries different possible hyperplanes and selects the one that:
  - Maximizes this margin  
  - Ensures the best separation between classes  

##### Key Idea

- The optimal hyperplane is the one that creates the **maximum distance between π⁺ and π⁻**

##### Concept of Support Vectors

- Start by drawing a separating hyperplane that divides the data into two classes  

- From this hyperplane, draw two parallel boundaries:
  - One towards the positive class → π⁺  
  - One towards the negative class → π⁻  

- Move these planes outward until they touch the nearest data points from each class  

- The first data points that touch:
  - π⁺ (positive side)  
  - π⁻ (negative side)  
  are called **Support Vectors**  

##### Why Support Vectors are Important

- These are the **closest points to the decision boundary**  
- They define the **margin of the classifier**  
- The position of the hyperplane depends only on these points  
- Even if other points are removed, the boundary remains the same (as long as support vectors are  

##### Key Insight

- SVM does not care about all data points  
- It focuses only on the most critical ones → **support vectors**

##### Important Points about SVM

- SVMs are very robust to outliers  
  - Because only support vectors influence the decision boundary  
  - Far-away points have little to no effect  

- Can also work on non-linear data (using kernels)  
  - Maps data to higher dimensions where it becomes separable  
  - Common kernels: Linear, Polynomial, RBF  

- SVM can be implemented for both:
  - Classification tasks  
  - Regression tasks (SVR - Support Vector Regression)
