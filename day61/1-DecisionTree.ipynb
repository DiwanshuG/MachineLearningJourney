### Geometric Intuition

##### Geometric Intuition of Decision Trees

##### Feature Space View
In a 2D dataset:
- X-axis → Feature 1 (e.g., Petal Length)
- Y-axis → Feature 2 (e.g., Sepal Length)

Each data point represents a coordinate in feature space.

---

##### How Decision Trees Split Space
Decision Trees split using threshold rules:

Feature ≤ threshold  
Feature > threshold  

Geometrically:
- Splitting on Feature 1 creates a vertical line.
- Splitting on Feature 2 creates a horizontal line.

---

##### Axis-Aligned Splits
Decision Trees create axis-aligned decision boundaries.
- No diagonal splits
- Only vertical or horizontal cuts
- Recursive partitioning of space

---

##### Recursive Space Partitioning
Algorithm intuition:
1. Choose the best feature.
2. Choose the best threshold.
3. Split space into two regions.
4. Repeat inside each region.

Each split creates smaller rectangular regions.

---

##### Decision Regions
- Feature space gets divided into rectangles.
- Each rectangle predicts one class.
- Each rectangle corresponds to a leaf node.
- The boundary appears step-wise (box-like).

---

##### Comparison with Logistic Regression

Logistic Regression:
- Produces a single straight line.
- Linear decision boundary.
- Smooth separation.

Decision Tree:
- Produces step-like boundaries.
- Piecewise constant regions.
- Overall non-linear behavior.

---

##### Overfitting Intuition
Deep trees:
- Create many small rectangles.
- Fit noise in training data.
- Produce jagged decision boundaries.
- Lead to high variance.

---


A Decision Tree performs recursive axis-aligned partitioning of feature space.

It does not learn a single equation.  
It learns regions.

## Decision Tree Pseudo Code

1. Begin with the training dataset, which contains:
   - Feature variables (X)
   - Target variable (y) for classification or regression

2. If all examples belong to the same class:
   - Create a leaf node with that class.
   - Stop.

3. If no features remain or stopping criteria is met:
   - Create a leaf node with the majority class (classification)
     or mean value (regression).
   - Stop.

4. Determine the best feature to split on:
   - For classification → use Information Gain or Gini Index.
   - For regression → use variance reduction (MSE reduction).

5. Find the best threshold (for numerical features) 
   or best category split (for categorical features).

6. Split the dataset into subsets based on the selected feature and threshold.
   - Each split creates a decision node in the tree.

7. Recursively repeat the process for each subset:
   - Treat each subset as a new dataset.
   - Apply steps 2–6.

8. Continue until:
   - Nodes are pure, or
   - Maximum depth is reached, or
   - Minimum samples per node condition is satisfied.

The final result is a tree where:
- Internal nodes represent feature-based decisions.
- Branches represent outcomes of those decisions.
- Leaf nodes represent final predictions.

##### Programmatic Interpretation

Programmatically, a Decision Tree is a large structure of nested if-else conditions.

Each internal node represents a condition:
Feature ≤ threshold

Each branch represents the outcome of that condition (True / False).

Each leaf node represents the final prediction.

So in code form, a Decision Tree is simply recursive nested conditional logic.

---

##### Mathematical Interpretation

Mathematically, Decision Trees split the feature space using hyperplanes.

In 2D:
- A split is a line.
- The line is parallel to one of the axes.

In 3D:
- A split is a plane parallel to one coordinate axis.

In higher dimensions:
- A split is a hyperplane parallel to one of the feature axes.

---

##### Geometric Consequence

Because splits are always axis-aligned:

- The feature space gets divided into rectangular regions (2D).
- In higher dimensions, these regions are hyper-rectangles (also called hyper-cuboids).

Each region corresponds to one leaf node and predicts a constant value or class.

---



Decision Trees:
- Do not create diagonal boundaries.
- Partition space using axis-parallel hyperplanes.
- Form piecewise constant prediction regions.

> mathematically speaking , decision tree use hyperplanes which run parallel to any one of the axes to cut your coordiante system into hyper cuboids

#### Root Node
The topmost node of a Decision Tree.  
It represents the entire dataset and is the first feature selected for splitting based on the best criterion (e.g., highest Information Gain or lowest Gini).

---

#### Decision Node
An internal node where a feature-based decision is made.  
It contains a condition such as:

Feature ≤ threshold  

or  

Feature = category  

Each decision node splits the data into smaller subsets.

---

#### Leaf Node
The terminal node of the tree.  
It does not split further and represents the final prediction:
- Majority class (classification)
- Mean/median value (regression)

---

#### Splitting
The process of dividing the dataset into subsets based on a selected feature and threshold.  
Splitting is done to increase purity (classification) or reduce variance (regression).

---

#### Branch
A connection between two nodes.  
It represents the outcome of a decision condition (e.g., True/False or Yes/No) and leads to the next node in the tree.

# Decision Tree Entropy 

#### Entropy in Decision Trees

Entropy is a measure of impurity or randomness in a dataset.

It tells us how mixed the classes are in a node.

If all samples belong to one class → Entropy = 0 (pure node)  
If classes are evenly mixed → Entropy is maximum (high impurity)

---

#### Entropy Formula

For a classification problem:
$$
H(S) = - ∑ (pᵢ log₂ pᵢ)
$$
Where:
- pᵢ = probability of class i
- Log base 2 is used

---

#### Understanding with Binary Classification

If a node has:

- p = probability of Yes
- q = probability of No

Then:

$$
H(S) = - (p log₂ p + q log₂ q)
$$

---

#### Special Cases

1. Pure Node  
   Example: 10 Yes, 0 No  
   H = 0  

2. Completely Mixed (50-50)  
   Example: 5 Yes, 5 No  
   H = 1 (maximum for binary case)

3. Uneven Split  
   Example: 8 Yes, 2 No  
   Entropy is between 0 and 1

---

#### Why Entropy is Used

Decision Trees use Entropy to decide:
Which feature gives the best split?

The goal is to reduce entropy after splitting.

---

#### Information Gain

Information Gain measures how much entropy decreases after a split.

IG = Entropy(parent) − Weighted Entropy(children)

The feature with the highest Information Gain is selected for splitting.

---

High entropy → data is mixed → bad  
Low entropy → data is pure → good  

A Decision Tree tries to make nodes purer at every split.

##### more knowledge => less entropy

#### Example: Calculating Entropy Step-by-Step

Suppose we have a dataset:

Total samples = 10  
- Yes = 6  
- No = 4  

We want to calculate the Entropy of this dataset.

---

#### Step 1: Compute Probabilities

p(Yes) = 6 / 10 = 0.6  
p(No)  = 4 / 10 = 0.4  

---

#### Step 2: Apply Entropy Formula

H(S) = - [ p(Yes) log₂ p(Yes) + p(No) log₂ p(No) ]

H(S) = - [ (0.6 log₂ 0.6) + (0.4 log₂ 0.4) ]

---

#### Step 3: Calculate Log Values

log₂(0.6) ≈ -0.737  
log₂(0.4) ≈ -1.322  

Now substitute:

H(S) = - [ (0.6 × -0.737) + (0.4 × -1.322) ]

H(S) = - [ -0.4422 + (-0.5288) ]

H(S) = - ( -0.971 )

H(S) ≈ 0.971

---

#### Final Answer

Entropy ≈ 0.97

This means:
- The dataset is quite mixed.
- Not perfectly pure.
- Not completely random (which would be 1 in binary case).

---

#### Compare With Pure Case

If:
- Yes = 10
- No = 0

Then:

H(S) = - [ 1 log₂ 1 + 0 log₂ 0 ]
     = - [ 0 + 0 ]
     = 0

Pure node → Entropy = 0

#### Entropy vs Probability

Entropy measures the impurity or uncertainty of a dataset.
It depends directly on the class probability distribution.

For binary classification:

$$
H(p) = - \big( p \log_2 p + (1 - p) \log_2 (1 - p) \big)
$$

Where:
- p = probability of class 1
- (1 − p) = probability of class 2

---

#### Key Observations

1. When p = 0 or p = 1  
   $$ H(p) = 0 $$  
   The node is pure (no uncertainty).

2. When p = 0.5  
   $$ H(p) = 1 $$  
   Entropy is maximum (maximum uncertainty).

3. When p is near 0 or near 1  
   Entropy is low.

4. When p is near 0.5  
   Entropy is high.

---

#### Intuition

- If one class dominates → low entropy.
- If classes are evenly mixed → high entropy.
- Entropy is symmetric around p = 0.5.
- Entropy curve is concave (inverted U-shape).

$$
\text{Information Gain} = E(\text{Parent}) - E_{\text{children}}
$$

Substitute the calculated values:

\text{IG} = 0.97 - 0.69

$$

$$

#### Step 5: Calculate Information Gain for All Columns

Compute the Information Gain for each feature (column) in the dataset:

$$
\text{IG}(\text{Feature}) = E(\text{Parent}) - E_{\text{children (Feature)}}
$$

Repeat this process for every feature.

---

The feature with the highest Information Gain
(i.e., maximum decrease in entropy)
is selected as the splitting feature.

In other words,

The algorithm chooses the column that produces
the purest child nodes after the split.

#### Step 6: Find Information Gain Recursively

The Decision Tree applies a recursive greedy search algorithm
in a top-down manner to compute Information Gain at every level of the tree.

At each node:

1. Calculate the entropy of the current dataset.
2. Compute Information Gain for all available features.
3. Select the feature with the highest Information Gain.
4. Split the dataset based on that feature.
5. Repeat the process for each resulting child node.

This process continues recursively until:

- A leaf node is reached (Entropy = 0), or
- No further useful splits are possible.

Once a leaf node is reached, no more splitting is performed.
The leaf node represents the final prediction.

# Gini Impurity 

#### Entropy Formula for Two Classes (Binary Classification)

For a binary classification problem with two classes:

- p = probability of class 1  
- (1 − p) = probability of class 2  

The entropy formula is:

$$
H(p) = - \left( p \log_2 p + (1 - p) \log_2 (1 - p) \right)
$$

---

#### Alternative Notation

If:
- p₁ = probability of class 1
- p₂ = probability of class 2

Then:

$$
H = - (p_1 \log_2 p_1 + p_2 \log_2 p_2)
$$

---

#### Special Cases

- If p = 0 or p = 1  
  $$ H = 0 $$  
  (Pure node)

- If p = 0.5  
  $$ H = 1 $$  
  (Maximum entropy in binary case)

---
Entropy measures uncertainty.

It is:
- Minimum (0) when the node is pure.
- Maximum (1) when both classes are equally likely.

#### Gini Impurity Formula

For a node with K classes, the Gini Impurity is defined as:

$$
Gini = 1 - \sum_{i=1}^{K} p_i^2
$$

Where:
- \( p_i \) = probability of class i in the node

---

#### Binary Classification Case

If:
- p = probability of class 1
- (1 − p) = probability of class 2

Then:

$$
Gini = 1 - \left( p^2 + (1 - p)^2 \right)
$$

It can also be simplified to:

$$
Gini = 2p(1 - p)
$$

---

#### Special Cases

- If p = 0 or p = 1  
  $$ Gini = 0 $$  
  (Pure node)

- If p = 0.5  
  $$ Gini = 0.5 $$  
  (Maximum impurity for binary case)

---

#### Key Idea

Lower Gini → More pure node  
Higher Gini → More mixed node  

Decision Trees choose splits that minimize the weighted Gini impurity.

#### Difference Between Entropy and Gini Impurity

| Aspect | Entropy | Gini Impurity |
|--------|----------|---------------|
| Formula | $$ H = - \sum p_i \log_2 p_i $$ | $$ Gini = 1 - \sum p_i^2 $$ |
| Binary Formula | $$ H(p) = - \left( p \log_2 p + (1-p) \log_2 (1-p) \right) $$ | $$ Gini = 1 - \left( p^2 + (1-p)^2 \right) $$ |
| Range (Binary) | $$ 0 \le H \le 1 $$ | $$ 0 \le Gini \le 0.5 $$ |
| Pure Node Value | $$ H = 0 $$ | $$ Gini = 0 $$ |
| Maximum (Binary) | $$ H = 1 \text{ at } p=0.5 $$ | $$ Gini = 0.5 \text{ at } p=0.5 $$ |
| Interpretation | Measures uncertainty (information theory) | Measures probability of misclassification |
| Computational Cost | Uses $$ \log $$ (slightly slower) | No logarithm (faster) |
| Common Algorithm | Used in ID3 | Used in CART (default in sklearn) |

- Low Gini → Pure

#### Gini Impurity vs Entropy in Practice

In large datasets, Gini Impurity is often preferred over Entropy because:

- Gini does not use logarithms.
- It only requires squaring probabilities.
- Computationally, squaring is faster than computing logarithms.
- Therefore, Gini is slightly more efficient for large-scale training.

However, in some specific cases:

- Entropy (Information Gain) may produce more balanced splits.
- It can sometimes create more informative or slightly better-structured trees.
- Entropy penalizes impurity more strongly than Gini.

---

#### Practical Insight

- Gini is generally faster and is the default in many libraries (e.g., CART, sklearn).
- Entropy may perform better in certain datasets depending on class distribution.
- In practice, it can be useful to try both and compare results.

Both metrics aim to reduce impurity — they just measure it differently.

##### Splitting Numerical Feature Using Thresholds

For a numerical feature, the dataset is split using different threshold values.

Let:

- D = Complete dataset  
- f = Numerical feature (e.g., User Rating)  
- v1, v2, v3, ..., vn = Possible values of feature f  

For each possible threshold value, we try:

D → (f > v1) → D1, D2  
D → (f > v2) → D1, D2  
D → (f > v3) → D1, D2  
...  
D → (f > vn) → D1, D2  

Each condition divides the dataset into:

- D1 = Records satisfying the condition  
- D2 = Records not satisfying the condition  

For every possible threshold:
- Compute impurity (Entropy or Gini)
- Calculate weighted impurity
- Select the threshold that gives the best split (minimum impurity)

This is how decision trees handle numerical features.

##### Computing Information Gain for All Possible Thresholds

Let:

- D = Complete dataset  
- f = Numerical feature  
- v1, v2, v3, ..., vn = Candidate threshold values  
- D1, D2 = Subsets after split  
- E1, E2 = Impurity of D1 and D2  
- WEi = Weighted impurity for threshold vi  
- IGi = Information Gain for threshold vi  

---

For each threshold vi:

1. Split the dataset:

   D → (f > vi) → D1, D2  

2. Compute impurity of subsets:

   E1 = Impurity(D1)  
   E2 = Impurity(D2)  

3. Compute weighted impurity:

   WEi = (|D1| / |D|) * E1 + (|D2| / |D|) * E2  

4. Compute Information Gain:

   IGi = E(D) − WEi  

---

This process is repeated for:

- f > v1 → WE1 → IG1  
- f > v2 → WE2 → IG2  
- f > v3 → WE3 → IG3  
- ...  
- f > vn → WEn → IGn  

Finally:

Select the threshold vi that gives the maximum Information Gain (IGi).

That threshold becomes the best split for the numerical feature.
