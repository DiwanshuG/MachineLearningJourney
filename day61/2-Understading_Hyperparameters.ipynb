> #### overfitting 

##### Overfitting in Decision Trees

When a decision tree is allowed to grow without restriction 
(e.g., max_depth = None), it can become very deep and complex.

Such a tree:

- Performs very well on training data.
- Performs poorly on test data.
- Memorizes noise and outliers.

---

##### What Causes Overfitting?

If the tree keeps splitting until nodes become pure:

- It may create very small regions.
- It may split based on noisy, outlier, or erroneous data.
- It captures random fluctuations instead of true patterns.

Example:

A deep branch may isolate a single data point 
just to reduce impurity, even if it represents noise.

---

##### Effect of max_depth = None

- No restriction on tree depth.
- Tree keeps growing until all leaves are pure 
  or minimum samples condition is met.
- Training accuracy becomes very high.
- Generalization performance decreases.

---



Deep tree → Low bias, High variance  
Shallow tree → Higher bias, Lower variance  

Controlling max_depth helps prevent overfitting.

> ### underfitting

Underfitting occurs when a decision tree is too simple 
to capture the underlying pattern in the data.

This usually happens when:

- max_depth is too small
- Too few splits are allowed
- Strong regularization is applied

---

##### What Happens in Underfitting?

- The tree makes very few splits.
- Large regions are created.
- The model fails to capture important relationships.
- Both training and test accuracy are low.

---

##### Example

If max_depth = 1:

- The tree becomes a decision stump.
- Only one split is made.
- Complex patterns cannot be learned.

---

##### Effect on Bias and Variance

Underfitting means:

- High bias
- Low variance

The model is too rigid and makes strong assumptions.

---
Shallow tree → High bias → Underfitting  
Deep tree → High variance → Overfitting  

The goal is to choose a depth that balances both.

> ### 1) criteria 

##### 1. criterion

The `criterion` hyperparameter decides how the quality of a split is measured.

It defines the impurity metric used to choose the best split at each node.

---

##### Available Options (Classification)

1. "gini"  
   Uses Gini Impurity.

   Gini = 1 - ∑ p_i²

2. "entropy"  
   Uses Information Gain (based on Entropy).

   H = - ∑ p_i log₂ p_i

3. "log_loss" (newer versions of sklearn)  
   Similar to entropy but based on log loss formulation.

---

##### What Does It Control?

At every split:

- The tree evaluates all possible splits.
- It computes impurity using the chosen criterion.
- It selects the split that minimizes impurity (or maximizes information gain).

---

##### Gini vs Entropy

Gini:
- Faster (no logarithm).
- Default in sklearn.
- Often produces similar results.

Entropy:
- Slightly slower (uses log).
- More sensitive to class imbalance.
- Sometimes gives more balanced splits.

---

##### Does It Change the Tree Much?

In most practical cases:
- The difference is small.
- Both produce similar decision boundaries.

But:
- On imbalanced datasets, entropy may behave slightly differently.

---

##### When to Use What?

Default → "gini"  
If experimenting → try "entropy"  
If performance difference matters → compare both.

> ### 2) max_depth

max_depth controls the maximum number of levels 
the decision tree is allowed to grow.

It limits how deep the tree can become.

---

##### What Is Depth?

Depth = number of splits from root to deepest leaf.

Example:

Root → Level 0  
Children → Level 1  
Grandchildren → Level 2  

If max_depth = 2,
the tree cannot grow beyond level 2.

---

##### What Happens When max_depth Is Small?

Example: max_depth = 1 or 2

- Very few splits
- Large decision regions
- Simple model
- High bias
- Low variance
- Risk of underfitting

---

##### What Happens When max_depth Is Large?

Example: max_depth = None

- Tree grows until pure
- Very small regions
- Complex structure
- Low bias
- High variance
- Risk of overfitting

---

##### Extreme Cases

max_depth = 1  
→ Decision stump (only one split)

max_depth = None  
→ Tree grows until:
   - All leaves are pure
   - Or other stopping criteria apply

---

##### Why It Is Important

max_depth is one of the strongest regularization controls.

It directly controls model complexity.

Small depth → simpler model  
Large depth → complex model  

---

##### Bias-Variance Tradeoff

Small max_depth → High bias, Low variance  
Large max_depth → Low bias, High variance  

The goal is to find a balanced depth.

---

##### Practical Tip

Always tune max_depth using:
- Cross-validation
- Grid search

It often has a bigger impact than changing the criterion.

> ### 3) min_samples_split: int or float, default=2

min_samples_split defines the minimum number of samples 
required at a node to allow further splitting.

If a node has fewer samples than this value,
the node becomes a leaf.

---

##### If Integer

Example:

min_samples_split = 5

A node must have at least 5 samples
to be eligible for splitting.

If it has 4 or fewer → no split allowed.

---

##### If Float

When given as a float:

min_samples_split = fraction of total samples

Example:

min_samples_split = 0.1  
Dataset size = 1000  

Minimum samples required = 0.1 × 1000 = 100

This makes it scale automatically with dataset size.

---

##### Why It Is Important

It prevents the tree from:

- Splitting very small nodes
- Creating tiny branches
- Overfitting on noise

---

##### Effect of Small Value (e.g., 2)

- Tree splits aggressively
- Can create many small nodes
- High variance
- Risk of overfitting

---

##### Effect of Large Value

- Fewer splits
- Simpler tree
- Higher bias
- Risk of underfitting

---

##### Important Difference from max_depth

max_depth limits vertical growth (height of tree)

min_samples_split limits horizontal growth 
(number of splits allowed based on data size)

---

##### Practical Insight

Default = 2  
Good for most cases.

For noisy data:
Increase it slightly (e.g., 5, 10, or 0.01).

It acts as a regularization parameter.

> ### 4) min_samples_leaf: int or float, default=1

min_samples_leaf defines the minimum number of samples 
required to be present in each leaf node.

If after a split, any child node has fewer samples 
than this value, the split is not allowed.

---

##### If Integer

Example:

min_samples_leaf = 5

Every leaf must contain at least 5 samples.

If a split produces a leaf with 3 samples,
that split will be rejected.

---

##### If Float

When given as a float:

min_samples_leaf = fraction of total samples

Example:

min_samples_leaf = 0.05  
Dataset size = 1000  

Minimum samples per leaf = 0.05 × 1000 = 50

Each leaf must contain at least 50 samples.

---

##### Why It Is Important

It prevents:

- Very small leaf nodes
- Splits that isolate single samples
- Learning from noise
- Overfitting

It smooths the decision boundary.

---

##### Effect of Small Value (default = 1)

- Leaves can contain 1 sample
- Very flexible model
- Low bias
- High variance
- High risk of overfitting

---

##### Effect of Larger Value

- Larger leaf regions
- Simpler decision boundary
- Higher bias
- Lower variance
- Reduced overfitting

---

##### Difference Between min_samples_split and min_samples_leaf

min_samples_split:
- Controls when a node can split.

min_samples_leaf:
- Controls how small a final leaf can be.

min_samples_leaf is usually stronger regularization 
than min_samples_split.

---


If your tree overfits:

Increase min_samples_leaf first.

It often gives smoother and more stable trees.

> ### 5. max_features (int, float, {"sqrt", "log2"}, default=None)

max_features controls how many features 
are considered when looking for the best split.

Instead of checking all features at every node,
the algorithm randomly selects a subset of features.

The best split is chosen only from that subset.

---

##### Default (None)

All features are considered at every split.

If dataset has m features:
All m features are evaluated.

This is standard Decision Tree behavior.

---

##### If Integer

Example:

max_features = 3

At every split:
Only 3 randomly selected features are considered.

Even if dataset has 20 features,
only 3 are evaluated per split.

---

##### If Float

Example:

max_features = 0.5

Means:
Use 50% of total features at each split.

If dataset has 10 features:
5 features will be randomly selected.

---

##### If "sqrt"

Use:

sqrt(total_features)

Commonly used in classification.

Example:
If 16 features → sqrt(16) = 4 features per split.

---

##### If "log2"

Use:

log2(total_features)

If 16 features → log2(16) = 4 features per split.

---

##### Why It Is Important

Reducing max_features:

- Adds randomness
- Reduces correlation between splits
- Reduces variance
- Prevents dominance of strong features

---

##### Effect of Large Value (None)

- Deterministic tree
- More greedy
- Lower bias
- Higher variance

---

##### Effect of Smaller Value

- More randomness
- Slightly higher bias
- Lower variance
- More robust model

---

##### Very Important Insight

In a single Decision Tree:
max_features usually not critical.

In Random Forest:
max_features is extremely important.

It creates diversity among trees.

---

##### Practical Defaults

Classification:
max_features = "sqrt"

Regression:
max_features = None or 1.0

---

This hyperparameter controls feature-level randomness.
