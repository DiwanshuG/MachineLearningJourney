##### Before Minimizing the Loss Function  

Before finding the minimum of the cross entropy loss function,  
we first need to understand how to compute the derivative of the sigmoid function.

Since logistic regression uses:

$$
\hat{y} = \sigma(z)
$$

where

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

we must compute:

$$
\frac{d}{dz} \sigma(z)
$$

$$
\frac{d}{dx} \left( \frac{1}{1 + e^{-x}} \right)
=
\frac{e^{-x}}{(1 + e^{-x})^2}
$$

Now rewrite the denominator:

$$
(1 + e^{-x})(1 + e^{-x})
$$

So,

$$
=
\frac{1}{1 + e^{-x}} \cdot \frac{e^{-x}}{1 + e^{-x}}
$$

Notice that:

$$
\frac{1}{1 + e^{-x}} = \sigma(x)
$$

and

$$
\frac{e^{-x}}{1 + e^{-x}} = 1 - \sigma(x)
$$

Therefore,

$$
\frac{d}{dx} \sigma(x)
=
\sigma(x)\big(1 - \sigma(x)\big)
$$
