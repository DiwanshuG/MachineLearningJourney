```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_pos = np.array([[3,5], [5,4], [4,6]])   # green points
X_neg = np.array([[1,3], [2,1], [3,2]])   # red points
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

ax = axes[0]

ax.scatter(X_pos[:,0], X_pos[:,1], color='green', s=100)
ax.scatter(X_neg[:,0], X_neg[:,1], color='red', s=100)

# Line: y = 0.3x + 2
x = np.linspace(0, 6, 100)
y = 0.3*x + 2

ax.plot(x, y, color='white', linewidth=2)

ax.set_title("Model 1")
ax.set_xlim(0,6)
ax.set_ylim(0,7)

ax.text(4.5,6.2,"+ve region", color='green')
ax.text(0.5,0.5,"-ve region", color='red')

ax = axes[1]

ax.scatter(X_pos[:,0], X_pos[:,1], color='green', s=100)
ax.scatter(X_neg[:,0], X_neg[:,1], color='red', s=100)

# Line: y = -1.2x + 6
y2 = -1.2*x + 6

ax.plot(x, y2, color='white', linewidth=2)

ax.set_title("Model 2")
ax.set_xlim(0,6)
ax.set_ylim(0,7)

ax.text(4.5,6.2,"+ve region", color='green')
ax.text(3.5,0.5,"-ve region", color='red')

# Styling
for ax in axes:
    ax.set_facecolor('black')
    ax.tick_params(colors='white')
    for spine in ax.spines.values():
        spine.set_color('white')

plt.tight_layout()
plt.show()
```

> above is wrong 

the correct formula will be : 
$$
L = - y_i \log(\hat{y}_i) - (1 - y_i)\log(1 - \hat{y}_i)
$$

$$
L = \sum_{i=1}^{n} 
\left[
- y_i \log(\hat{y}_i) 
- (1 - y_i)\log(1 - \hat{y}_i)
\right]
$$

Log-loss error  OR Binary Cross Entropy 
$$
L = -\frac{1}{n} \sum_{i=1}^{n}
\left[
y_i \log(\hat{y}_i) 
+ (1 - y_i)\log(1 - \hat{y}_i)
\right]
$$

There is one way to find the minmum of above funtion : **Gradient Descent**
