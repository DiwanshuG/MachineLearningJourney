Log-loss error  OR Binary Cross Entropy 
$$
L = -\frac{1}{n} \sum_{i=1}^{n}
\left[
y_i \log(\hat{y}_i) 
+ (1 - y_i)\log(1 - \hat{y}_i)
\right]
$$

##### Dataset Representation  

Rows = $m$  
Columns = $n$  

We represent the dataset as a matrix:

$$
X =
\begin{bmatrix}
x_{11} & x_{12} & x_{13} & \dots & x_{1n} \\
x_{21} & x_{22} & x_{23} & \dots & x_{2n} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
x_{m1} & x_{m2} & x_{m3} & \dots & x_{mn}
\end{bmatrix}
$$

Where:

- $m$ = number of data points (rows)
- $n$ = number of features (columns)
- $x_{ij}$ = value of the $j^{th}$ feature for the $i^{th}$ data point

---

##### Target Vector  

The corresponding output labels are:

$$
Y =
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_m
\end{bmatrix}
$$

Each row of $X$ corresponds to one $y_i$ in $Y$.

##### Vector Form of Predictions in Logistic Regression  

The predicted outputs for all $m$ data points can be written as:

$$
\hat{Y} =
\begin{bmatrix}
\hat{y}_1 \\
\hat{y}_2 \\
\vdots \\
\hat{y}_m
\end{bmatrix}
$$

where each prediction is obtained using the sigmoid function:

$$
\hat{y}_1 = \sigma(w_0 + w_1 x_{11} + w_2 x_{12} + \dots + w_n x_{1n})
$$

$$
\hat{y}_2 = \sigma(w_0 + w_1 x_{21} + w_2 x_{22} + \dots + w_n x_{2n})
$$

$$
\vdots
$$

$$
\hat{y}_m = \sigma(w_0 + w_1 x_{m1} + w_2 x_{m2} + \dots + w_n x_{mn})
$$

---

##### Compact Matrix Form  

This can be written compactly as:

$$
\hat{Y} = \sigma(XW)
$$

where:

- $X$ is the $m \times n$ feature matrix  
- $W$ is the weight vector  
- $\sigma(\cdot)$ is applied element-wise  

Log-loss error  OR Binary Cross Entropy 
$$
L = -\frac{1}{m} \sum_{i=1}^{m}
\left[
y_i \log(\hat{y}_i) 
+ (1 - y_i)\log(1 - \hat{y}_i)
\right]
$$



$$
\sum_{i=1}^{m} y_i \log(\hat{y}_i)
=
y_1 \log(\hat{y}_1)
+ y_2 \log(\hat{y}_2)
+ y_3 \log(\hat{y}_3)
+ \dots
+ y_m \log(\hat{y}_m)
$$

$$
\begin{bmatrix}
y_1 & y_2 & y_3 & \dots & y_m
\end{bmatrix}
\;
\log
\left(
\begin{bmatrix}
\hat{y}_1 \\
\hat{y}_2 \\
\hat{y}_3 \\
\vdots \\
\hat{y}_m
\end{bmatrix}
\right)
$$

$$
Y \log(\hat{Y}) = Y \log\big(\sigma(XW)\big)
$$

so the loss function becomes : 
$$
L = -\frac{1}{m}
\left[
Y \log(\hat{Y}) + (1 - Y)\log(1 - \hat{Y})
\right]
$$

where

$$
\hat{Y} = \sigma(XW)
$$

##### Gradient Descent Update Rule  

For each epoch:

$$
W = W - \eta \frac{\partial L}{\partial W}
$$

$$
L = -\frac{1}{m}
\left[
y \log(\hat{y}) + (1 - y)\log(1 - \hat{y})
\right]
$$

$$
\frac{\partial L}{\partial W}
=
\left[
\frac{\partial L}{\partial w_0},
\frac{\partial L}{\partial w_1},
\frac{\partial L}{\partial w_2},
\;\dots,\;
\frac{\partial L}{\partial w_n}
\right]
$$

##### Derivative of the First Term in Cross Entropy  

We compute:

$$
\frac{d}{dW} \left( y \log(\hat{y}) \right)
$$

Since $y$ is constant with respect to $W$:

$$
= y \frac{d}{dW} \left( \log(\hat{y}) \right)
$$

Using chain rule:

$$
= y \cdot \frac{1}{\hat{y}} \cdot \frac{d\hat{y}}{dW}
$$

---

##### Derivative of Sigmoid  

We know:

$$
\hat{y} = \sigma(WX)
$$

and

$$
\frac{d}{dz} \sigma(z) = \sigma(z)(1 - \sigma(z))
$$

So,

$$
\frac{d\hat{y}}{dW}
=
\sigma(WX)(1 - \sigma(WX)) \cdot \frac{d(WX)}{dW}
$$

Since:

$$
\frac{d(WX)}{dW} = X
$$

we get:

$$
\frac{d\hat{y}}{dW}
=
\hat{y}(1 - \hat{y}) X
$$

---

##### Substituting Back  

$$
y \cdot \frac{1}{\hat{y}} \cdot \hat{y}(1 - \hat{y}) X
=
y(1 - \hat{y}) X
$$

$$
- (1 - y)\,\frac{\hat{y}(1 - \hat{y}) X}{1 - \hat{y}}
=
- (1 - y)\,\hat{y} X
$$

$$
\frac{dL}{dW}
=
-\frac{1}{m}
\left[
y(1 - \hat{y})X
-
\hat{y}(1 - y)X
\right]
$$

Factor out $X$:

$$
=
-\frac{1}{m}
\left[
y(1 - \hat{y})
-
\hat{y}(1 - y)
\right] X
$$

$$
\frac{\partial L}{\partial W}
=
-\frac{1}{m}(y - \hat{y})X
$$

$$
\boxed{
W = W + \eta \frac{1}{m}(y - \hat{y})X
}
$$

```python
from sklearn.datasets import make_classification
import numpy as np

np.random.seed(41)
```

```python
X, y = make_classification(
    n_samples=100,
    n_features=2,
    n_informative=1,
    n_redundant=0,
    n_classes=2,
    n_clusters_per_class=1,
    class_sep=20,
    hypercube=False,
    random_state=41
)
```

```python
import matplotlib.pyplot as plt

y_flat = y.ravel()

plt.figure()
plt.scatter(X[:,0], X[:,1], c=y_flat, cmap='winter')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Generated Classification Dataset")

plt.show()
```

```python
lor = LogisticRegression(
    solver="lbfgs",
    max_iter=1000,
    random_state=41
)

lor.fit(X, y.ravel())
```

```python
m1 = -(lor.coef_[0][0]/lor.coef_[0][1])
b1 = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m1*x_input + b1
```

```python
def gd(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.5
    
    for i in range(15000):
        y_hat = sigmoid(np.dot(X,weights))
        weights = weights + lr*(np.dot((y-y_hat),X)/X.shape[0])
        
    return weights[1:],weights[0]
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
coef_,intercept_ = gd(X,y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
```
