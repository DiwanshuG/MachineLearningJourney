## Bagging vs Boosting

---

## Overview

### Bagging (Bootstrap Aggregation)

- Used to reduce **variance**
- Works best with models having:
  - Low bias  
  - High variance  
- Example: Fully grown Decision Trees  

---

### Boosting

- Used to reduce **bias**
- Works best with models having:
  - High bias  
  - Low variance  
- Example: Decision Stumps  

---

## Key Differences

### 1. Type of Model Used

- **Bagging**:
  - Uses complex models  
  - Example: Deep Decision Trees  
  - Goal: reduce overfitting  

- **Boosting**:
  - Uses simple models  
  - Example: Decision Stumps  
  - Goal: improve weak learners  

---

### 2. Sequential vs Parallel Learning

- **Bagging**:
  - Models are trained **independently**
  - Training happens in **parallel**

- **Boosting**:
  - Models are trained **one after another**
  - Each model learns from previous mistakes  
  - Training is **sequential**

---

### 3. Weightage of Base Learners

- **Bagging**:
  - All models have **equal weight**
  - Final prediction = simple average / majority vote  

- **Boosting**:
  - Models have **different weights (alpha)**
  - Better models get more importance  

---

## Additional Differences

### Data Handling

- **Bagging**:
  - Uses random sampling with replacement  
  - Each model sees a different subset  

- **Boosting**:
  - Uses full dataset  
  - Adjusts weights to focus on hard examples  

---

### Error Handling

- **Bagging**:
  - No special focus on mistakes  

- **Boosting**:
  - Focuses more on misclassified points  

---

## Summary Table

| Feature | Bagging | Boosting |
|--------|--------|----------|
| Goal | Reduce variance | Reduce bias |
| Model Type | Complex (deep trees) | Simple (stumps) |
| Training | Parallel | Sequential |
| Weight of Models | Equal | Unequal |
| Focus on Errors | No | Yes |
| Data Sampling | Bootstrap sampling | Weighted data |

---

## Final Insight

- Bagging improves stability by averaging  
- Boosting improves accuracy by correcting mistakes  

Both are powerful ensemble techniques but solve different problems.
