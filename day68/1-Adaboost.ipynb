# Day 68 of Machine Learning: AdaBoost (Adaptive Boosting)

## 1. Introduction

AdaBoost is a classical machine learning algorithm introduced in 1995 by Yoav Freund and Robert Schapire. It is not part of deep learning and was developed much earlier.

Although newer algorithms like Gradient Boosting, XGBoost, and LightGBM often outperform AdaBoost today, it is still important to learn because it builds the foundation of boosting techniques.

---

## 2. Why Learn AdaBoost

- Forms the base concept for boosting methods
- Helps in understanding Gradient Boosting and advanced ensemble models
- Commonly asked in interviews for conceptual clarity

---

## 3. Weak Learners

A weak learner is a model that performs slightly better than random guessing.

- For binary classification:
  - Random accuracy = 50%
  - Weak learner accuracy ≈ 50–60%

AdaBoost combines multiple weak learners to create a strong model.

---

## 4. Decision Stumps

A decision stump is a decision tree with only one split (max_depth = 1).

### Example Dataset

| CGPA | IQ  | Placed |
|------|-----|--------|
| 8.5  | 110 | 1      |
| 6.2  | 95  | 0      |
| 7.8  | 105 | 1      |

### Decision Tree vs Decision Stump

- Decision Tree:
  - Multiple splits
  - High complexity

- Decision Stump:
  - Only one split
  - Very simple model

## 5. How the Best Split is Chosen

The split is selected based on maximum information gain (or minimum entropy).

Steps:
1. Try splitting on each feature (CGPA, IQ)
2. Calculate entropy reduction
3. Choose the split with highest information gain

---

## 6. Why Weak Learners Work in AdaBoost

- Each weak learner individually has low accuracy
- AdaBoost combines them sequentially
- Each new learner focuses more on previously misclassified points

This results in a strong final model.

---

## 7. AdaBoost Working (High-Level)

1. Initialize all data points with equal weights
2. Train first decision stump
3. Identify misclassified points
4. Increase weights of misclassified points
5. Train next stump focusing on difficult samples
6. Repeat for multiple iterations
7. Combine all models using weighted voting

---

## 8. Key Characteristics

- Sequential learning approach
- Focuses on hard-to-classify samples
- Uses weighted combination of models
- Typically uses decision stumps as base learners

---

## 9. Limitations

- Sensitive to noise and outliers
- Can overfit if too many iterations
- Less powerful compared to modern boosting algorithms

---

## 10. Modern Alternatives

- Gradient Boosting
- XGBoost
- LightGBM

These methods improve upon AdaBoost by handling noise better and providing higher performance.

## 12. Label Encoding in AdaBoost

In AdaBoost, class labels are typically encoded as:

- Positive class → $$+1$$  
- Negative class → $$-1$$  

### Why this encoding is used

This representation simplifies mathematical calculations, especially in:

- Weight updates  
- Final prediction (weighted sum of learners)

---

### Prediction Representation

Each weak learner predicts values from:

$$
\{+1, -1\}
$$

---

### Key Insight

- If prediction is correct:
  - The model reduces the importance (weight) of that data point  

- If prediction is wrong:
  - The model increases the importance (weight) of that data point  

---

### 
Using $$+1$$ and $$-1$$:
- Makes the algorithm mathematically simple  
- Helps in updating weights efficiently  
- Enables clean final decision using sign-based aggregation  

> ### Adaboost is a stage wise additive method 

## The Idea

We start by fitting the best decision stump (a tree with one split) that gives the maximum reduction in entropy.

Even after choosing the best split, some classification mistakes will occur.

---

### Step-by-Step Process

1. Train the first decision stump  
   - It makes some mistakes  
   - Assign a weight to this model → $$\alpha_1$$  

2. Increase focus on misclassified points  
   - This is similar to upsampling (giving more importance to गलत points)

3. Train the second decision stump  
   - It focuses more on previous mistakes  
   - It may correct earlier errors but can introduce new ones  
   - Assign weight → $$\alpha_2$$  

4. Repeat the process  
   - Again increase focus on misclassified points  
   - Train third stump  
   - Assign weight → $$\alpha_3$$  

---

### Final Model

After training multiple stumps, we combine them:

$$
h(x) = \text{sign} \big( \alpha_1 h_1(x) + \alpha_2 h_2(x) + \alpha_3 h_3(x) \big)
$$

---

### Example

Query point:
$$
x = (7.5, 81)
$$

Predictions from models:
- $$h_1(x) = -1$$  
- $$h_2(x) = +1$$  
- $$h_3(x) = -1$$  

Model weights:
- $$\alpha_1 = 2$$  
- $$\alpha_2 = 10$$  
- $$\alpha_3 = 1$$  

---

### Final Calculation

$$
\text{sign}(2 \cdot (-1) + 10 \cdot (1) + 1 \cdot (-1)) = \text{sign}(7) = +1
$$

---

### Conclusion

- Each model contributes based on its weight  
- Stronger models (higher $$\alpha$$) influence the final prediction more  
- Final output is determined using the sign of the weighted sum
